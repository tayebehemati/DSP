# مشخصات زمانی و فرکانسی سیگنال‌ها و سیستم‌ها

## ۶-۰ مقدمه  

توصیف یک سیستم LTI بر حسب پاسخ فرکانسی جایگزینی برای توصیف حوزه زمان سیستم بر حسب کانولوشن است.  
در تحلیل سیستم‌های LTI غالباً استفاده از حوزه فرکانسی مناسب‌تر است زیرا معادلات دیفرانسیل و تفاضلی و عمل کانولوشن در حوزه فرکانس به عملیات جبری تبدیل می‌شوند.  
همچنین مفاهیمی چون فرکانس‌گزینی را در حوزه فرکانس می‌توان راحت تصور کرد.  

البته در طراحی سیستم معمولاً باید هم ملاحظات حوزه فرکانس منظور شود و هم ملاحظات حوزه زمان.  
برای نمونه همان‌طور که در مثال‌های ۴-۱۸ و ۵-۱۲ به اختصار گفتیم، و در این فصل به تفصیل خواهیم گفت، رفتار نوسانی شدید پاسخ ضربه فیلترهای فرکانس‌گزین ممکن است نامطلوب باشد و مجبور شویم فرکانس‌گزینی را تا حدی فدای دستیابی به مشخصات مطلوب حوزه زمان کنیم.  

در عمل وضعیت‌های این‌چنین قاعده هستند نه استثنا، زیرا در اکثر کاربردها قیدهایی برای سیستم هم در حوزه زمان و هم در حوزه فرکانس تعیین می‌کنیم که به خواسته‌های متضاد می‌انجامد.  

پس در طراحی و تحلیل سیستم دانستن روابط بین حوزه فرکانس و حوزه زمان و بده‌بستان‌های موجود مهم است.  
معرفی این مباحث و روابط مهم‌ترین هدف این فصل است.  

## ۶-۱ نمایش اندازه - فاز تبدیل فوریه  

تبدیل فوریه در حالت کلی مختلط است و همان‌طور که قبلاً گفته‌ایم می‌توان آن را با بخش‌های حقیقی و موهومی یا اندازه و فازش نشان داد.  
نمایش اندازه - فاز تبدیل فوریه پیوسته در زمان \$X(j\omega)\$ عبارت است از:  

$$
X(j\omega) = |X(j\omega)| e^{j \angle X(j\omega)}
$$

(معادله ۶-۱)

به‌نحوی مشابه، نمایش اندازه - فاز تبدیل فوریه گسسته در زمان \$X(j\omega)\$ عبارت است از:

$$
X(j\omega) = |X(j\omega)| e^{j \angle X(j\omega)}
$$

(معادله ۶-۲)

در بحث زیر راجع به نمایش‌های اندازه - فاز توجه عمده خود را روی حالت پیوسته در زمان متمرکز می‌کنیم.  
نکات اصلی را به حالت گسسته در زمان نیز می‌توان اعمال کرد.  

از معادله ترکیب تبدیل فوریه یعنی معادله (۴-۸) چنین بر می‌آید که می‌توان \$X(j\omega)\$ را تجزیه‌ی سیگنال \$x(t)\$ به مجموع نمایی‌های مختلط با فرکانس‌های مختلف دانست.  

در حقیقت چنان‌که در بخش ۴-۳-۷ گفتیم \$|X(j\omega)|^2\$ را می‌توان طیف چگالی انرژی \$x(t)\$ دانست.  
یعنی \$|X(j\omega)|^2 \, d\omega / 2\pi\$ را می‌توان انرژی واقع در باند فرکانسی بی‌نهایت کوچک بین \$\omega\$ و \$\omega + d\omega\$ فرض کرد.  

پس اندازهٔ \$|X(j\omega)|\$ محتوای فرکانسی سیگنال را نشان می‌دهد.  
یعنی \$|X(j\omega)|\$ اطلاعات مربوط به اندازه نسبی نمایی‌های مختلط تشکیل‌دهندهٔ \$x(t)\$ را در بردارد.  

مثلاً اگر در خارج یک باند فرکانسی کوچک حول صفر داشته باشیم \$|X(j\omega)| = 0\$، آنگاه \$x(t)\$ تنها نوسان‌های فرکانس پایین دارد.

ولی زاویه فاز \$\angle X(j\omega)\$ بر اندازهٔ مولفه‌های فرکانسی اثری ندارد، بلکه اطلاعات مربوط به فاز نسبی این نمایی‌ها را نشان می‌دهد.  
رابطهٔ فاز موجود در \$\angle X(j\omega)\$ اثر زیادی بر طبیعت سیگنال \$x(t)\$ دارد و نوعاً اطلاعات زیادی در مورد سیگنال دارد.  

در واقع بسته به تابع فاز می‌توانیم سیگنال‌هایی با ظاهر متفاوت به دست آوریم، حتی اگر تابع اندازه تغییر نکند.  

### به عنوان نمونه:

دوباره مثال شکل ۳-۳ را در نظر بگیرید.  
در اینجا کشتی با جمع آثار سه موج روبرو است، که هر کدام با یک سیگنال سینوسی مدل شده‌اند.  

اگر این سه سینوسی دامنهٔ ثابتی داشته باشند، بسته به فاز نسبی‌شان می‌توانند دامنه بسیار بزرگ یا کوچکی داشته باشند.  
پس فازها برای کشتی کاملاً مهم‌اند.  

### مثال دیگر:

سیگنال زیر را در نظر بگیرید:  

$$
x(t) = 1 + \frac{1}{2} \cos(2\pi t + \varphi_1) + \cos(4\pi t + \varphi_2) + \frac{2}{3} \cos(6\pi t + \varphi_3)
$$

(معادله ۶-۳)

شکل ۳-۴ این سیگنال را به‌ازای \$\varphi_1 = \varphi_2 = \varphi_3 = 0\$ نشان می‌داد.  
در شکل ۱-۶ این \$x(t)\$ را به‌ازای چند فاز مخالف برای سینوس‌ها نشان داده‌ایم.  

همان‌طور که از شکل برمی‌آید، می‌بینیم که سیگنال حاصل بر حسب مقادیر مختلف فاز نسبی چقدر می‌تواند متفاوت باشد.  

به‌طور کلی تغییر تابع فاز \$X(j\omega)\$ به تغییر مشخصات حوزه زمان سیگنال \$x(t)\$ می‌انجامد.  

در بعضی موارد اعوجاج فاز مهم است و در بعضی موارد نه.  

مثلاً یکی از خواص سیستم شنوایی انسان عدم حساسیت به فاز است.  
اگر تبدیل فوریهٔ سیگنال یک صدای تنها (مثلاً یک مصوت) اعوجاج فاز داشته باشد ولی اندازه‌اش تغییر نکند،  
قابلیت فهم آن تغییر زیادی نمی‌کند، گرچه شکل موج آن در حوزه زمان می‌تواند شکلی کاملاً دگرگون به خود بگیرد.  

هرچند اعوجاج ملایم فاز، نظیر اعوجاج تک‌صداها، قابلیت تشخیص را از بین نمی‌برد،  
ولی اعوجاج فاز شدید گفتار، مسلماً قابلیت تشخیص آن‌ها را از بین می‌برد.


به عنوان مثالی بارز فرض کنید $x(t)$ سیگنال ضبط شده‌ی بیان یک جمله است. اگر نوار وارونه پخش شود، سیگنال $x(-t)$ به دست می‌آید، و اثر متناظر در حوزه فرکانس منفی شدن فاز است.  

$$
F\{x(-t)\} = x(-j\omega) = |X(j\omega)| e^{-j\angle x(j\omega)}
$$

شکل ۶-۱ سیگنال $x(t)$  

معادله (۶-۳) به ازای چند مقدار مختلف $\varphi_1$، $\varphi_2$، $\varphi_3$:  
الف) $\varphi_1 = \varphi_2 = \varphi_3 = 0$  
ب) $\varphi_1 = 12 \ rad$ ، $\varphi_2 = 8 \ rad$ ، $\varphi_3 = 4 \ rad$  
ج) $\varphi_1 = 6 \ rad$ ، $\varphi_2 = -2.7 \ rad$ ، $\varphi_3 = 0.93 \ rad$  
د) $\varphi_1 = 1.2 \ rad$ ، $\varphi_2 = 4.1 \ rad$ ، $\varphi_3 = -7.02 \ rad$  

یعنی اندازه طیف نوار وارونه با طیف سیگنال اصلی برابر است و تنها فاز آن تغییر کرده است. واضح است که این تغییر فاز اثر مهمی بر قابلیت فهم صدای ضبط شده دارد.  

بررسی تصاویر مثال دیگری است که اهمیت فاز را نشان می‌دهد. همانطور که در فصل ۳ به اختصار گفتیم، تصویر سیاه و سفید را می‌توان با سیگنال $x(t_1,t_2)$ نشان داد که $t_1$ مختصه‌ی افقی نقطه‌ای از تصویر، $t_2$ مختصه‌ی قائم آن نقطه و $x(t_1,t_2)$ روشنی تصویر در $(t_1,t_2)$ است.  

تبدیل فوریه $X(j\omega_1,j\omega_2)$ تصویر، تجزیه‌ی $x(t_1,t_2)$ به صورت نمایی‌های $e^{\omega_1 t_1}$ و $e^{\omega_2 t_2}$ را نشان می‌دهد که تغییرات مکان $x(t_1,t_2)$ در فرکانس‌های مختلف را در بردارد. جنبه‌های مقدماتی تحلیل فوریه دوبعدی در مسائل ۴-۵۳ و ۴-۵۶ بررسی شده‌اند.  

در دیدن تصاویر بخشی از مهم‌ترین اطلاعات در لبه‌ها و نواحی دارای کنتراست بالا نهفته است. نواحی دارای ماکزیمم و مینیمم روشنی در یک تصویر، محل‌هایی هستند که توابع نمایی مختلط دارای فرکانس‌های متفاوت هم‌فازند.  

بنابراین معقول است که تصور کنیم فاز تبدیل فوریه یک تصویر بخش اعظم اطلاعات راجع به تصویر را در بر داشته باشد و مخصوصاً اطلاعات راجع به لبه‌ها باید در فاز باشد.  

برای بررسی این تصور، شکل ۱-۴ را دوباره در شکل ۶-۲ (الف) نشان داده‌ایم.  
شکل ۶-۲ (ب) اندازه‌ی تبدیل فوریه را نشان می‌دهد.  

در این شکل محور افقی $\omega_1$ و محور عمودی $\omega_2$ و روشنی تصویر در نقطه $(\omega_1,\omega_2)$ متناسب با اندازه تبدیل $X(j\omega_1,j\omega_2)$ تصویر شکل ۶-۲ (الف) است.  

فاز این تبدیل هم به نحوی مشابه در شکل ۶-۲ (ج) نشان داده شده است.  

شکل ۶-۲ (د) نتیجه صفر کردن فاز $X(j\omega_1,j\omega_2)$، حفظ اندازه تبدیل و گرفتن عکس تبدیل را نشان می‌دهد.  

در شکل ۶-۲ (ه) اندازه $X(j\omega_1,j\omega_2)$ برابر یک قرار داده شده، ولی فاز حفظ شده است.  

سرانجام شکل ۶-۲ (و) تصویر حاصل از گرفتن عکس تبدیل از تابعی با فاز نشان داده شده در شکل ۶-۲ (ج) و اندازه تبدیل تصویری کاملاً متفاوت (یعنی تصویر شکل ۶-۲ (ز)) را نشان می‌دهد.  

این تصاویر به روشنی اهمیت فاز در نمایش تصاویر را نشان می‌دهند.

## ۶-۲ نمایش اندازه - فاز پاسخ فرکانسی سیستم‌های LTI  

طبق خاصیت کانولوشن تبدیل فوریه پیوسته در زمان، $Y(j\omega)$ خروجی یک سیستم LTI و تبدیل $X(j\omega)$ ورودی آن به صورت زیر به هم مربوط می‌شوند:  

$$
Y(j\omega) = H(j\omega)X(j\omega)
$$

که در آن $H(j\omega)$ پاسخ فرکانسی سیستم؛ یعنی تبدیل فوریه پاسخ ضربه سیستم است.  

به نحوی مشابه در حالت گسسته در زمان، تبدیل فوریه‌های ورودی $X(e^{jw})$ و خروجی $Y(e^{jw})$ سیستم LTI توسط پاسخ فرکانسی $H(e^{jw})$ به هم مربوط می‌شوند:  

$$
Y(e^{jw}) = H(e^{jw})X(e^{jw})
$$
(معادله ۶-۴)  

پس اثر سیستم LTI بر ورودی، تغییر اندازه مولفه‌های فرکانسی آن است.  

اگر از دید نمایش اندازه - فاز به این اثر بنگریم، جزئیات بیشتری را خواهیم دید.  

در حالت پیوسته در زمان:  

$$
|Y(j\omega)| = |H(j\omega)||X(j\omega)|
$$
(معادله ۶-۵)

و  

$$
\angle Y(j\omega) = \angle H(j\omega) + \angle X(j\omega)
$$
(معادله ۶-۶)

در حالت گسسته در زمان نیز رابطه دقیقاً مشابهی وجود دارد.  

معادله (۶-۵) نشان می‌دهد که اثر یک سیستم LTI بر اندازه تبدیل فوریه ورودی، تغییر دادن مقیاس آن بر اساس پاسخ فرکانسی است.  

به همین دلیل $|H(j\omega)|$ یا $|H(e^{jw})|$ را غالباً بهره سیستم می‌نامند.  

معادله (۶-۶) نشان می‌دهد که سیستم LTI فاز ورودی $\angle X(j\omega)$ را با افزودن فاز $\angle H(j\omega)$ به آن تغییر می‌دهد و به همین دلیل $\angle H(j\omega)$ را غالباً جابجایی فاز سیستم می‌نامند.  

جابجایی فاز سیستم می‌تواند رابطه نسبی فاز مولفه‌های ورودی را تغییر دهد و منجر به تغییر شدید چهره حوزه زمان ورودی شود، حتی در مواقعی که بهره سیستم در همه فرکانس‌ها ثابت است.  

تغییر اندازه و فاز توسط سیستم LTI می‌تواند مطلوب یا نامطلوب باشد، بسته به این‌که این تغییر چگونه صورت می‌گیرد.  

در حالت دوم، یعنی نامطلوب بودن تغییر، می‌گوییم دامنه و فاز **اعوجاج** یافته‌اند.  

در بخش‌های زیر مفاهیم و ابزارهایی برای ملموس کردن این اثرها پی‌ریزی خواهیم کرد.


## ۶-۲-۱ فاز خطی و غیر خطی

اگر تغییر فاز تابعی خطی از $\omega$ باشد، اثر آن تعبیر سرراستی در حوزه زمان دارد. سیستم LTI پیوسته در زمانی با پاسخ فرکانسی زیر را در نظر بگیرید:  

$$
H(j\omega) = e^{-j\omega t_0}
$$  

پس بهره سیستم یک و فاز آن خطی است، یعنی:  

$$
\angle H(j\omega) = -\omega t_0, \quad |H(j\omega)| = 1
$$  

در مثال ۴-۱۵ نشان دادیم که در سیستمی با این پاسخ فرکانسی، خروجی جابجا شده زمانی ورودی است، یعنی:  

$$
y(t) = x(t - t_0)
$$  

در حالت گسسته در زمان، فاز خطی اثری مشابه حالت پیوسته در زمان دارد، به شرطی که شیب فاز خطی عدد صحیح باشد. از مثال ۵-۱۱ به یاد داریم که سیستم LTI دارای پاسخ فرکانسی $e^{-j\omega n_0}$ تابع فاز خطی دارد و خروجی آن جابجا شده ورودی آن است، یعنی:  

$$
y[n] = x[n - n_0]
$$  

پس فاز خطی دارای شیب صحیح با جابجایی $x[n]$ به اندازه آن عدد صحیح متناظر است. اگر شیب فاز عدد صحیح نباشد، اثری که در حوزه زمان ایجاد می‌شود پیچیده‌تر است و در فصل ۷ بخش ۷-۵ مورد بحث قرار خواهد گرفت.  

به طور حسی می‌توان گفت که پوش مقادیر سیگنال جابجا می‌شود ولی خود مقادیر ممکن است تغییر کنند.  

تغییر فاز خطی به تغییر ساده و ملموسی در حوزه زمان منجر می‌شود، ولی اگر ورودی در معرض تغییر فاز غیر خطی قرار گیرد، مولفه‌های مختلف ورودی به نحوی تغییر می‌کنند که فازهای نسبی‌شان عوض می‌شود. با برنهادن این نمایی‌ها سیگنالی به دست می‌آید که منظری کاملاً متفاوت نسبت به سیگنال ورودی دارد.  

![شکل ۶-۳](https://raw.githubusercontent.com/tayebehemati/test/main/images/%D8%B4%DA%A9%D9%84%206-3.jpg)  

**شکل ۶-۳**  
(الف) سیگنال پیوسته در زمانی که به عنوان ورودی چند سیستم با پاسخ فرکانسی دارای بهره واحد به کار رفته است؛  
(ب) خروجی سیستم دارای فاز خطی؛  
(ج) خروجی سیستم دارای فاز غیر خطی؛  
(د) پاسخ سیستمی با همان تابع فاز سیستم (ج) به اضافه یک فاز خطی.  

سیستم دیگری با فاز غیر خطی که با افزودن یک فاز خطی به سیستم قبلی به دست آمده است، یعنی:  

$$
H_3(j\omega) = H_2(j\omega) \cdot e^{-j\omega n_0}
$$  

که در آن:  

$$
H_2(j\omega) = e^{j\angle H_2(j\omega)}
$$  

تابعی غیر خطی از $\omega$ است.  

پس خروجی شکل (د) را می‌توان پاسخ اتصال متوالی سیستم $H_2(j\omega)$ و سیستمی با فاز خطی دانست، به نحوی که سیگنال‌های شکل‌های (ج) و (د) تنها جابجا شده یکدیگر باشند.

**شکل ۶-۴**  
(الف) سیگنال گسسته در زمانی که به چند سیستم با اندازه پاسخ فرکانسی یک اعمال شده است؛  
(ب) پاسخ سیستم دارای فاز خطی با شیب ۵-؛  
(ج) پاسخ سیستم با فاز غیر خطی؛  
(د) پاسخ سیستمی که همان فاز غیر خطی شکل (ج) را دارد ولی یک فاز خطی نیز با آن جمع شده است.  

توجه کنید که تمام سیستم‌های در نظر گرفته شده در شکل‌های ۶-۳ و ۶-۴ بهره‌ای واحد دارند، یعنی اندازه تبدیل فوریه ورودی بدون تغییر در خروجی ظاهر می‌شود. به همین خاطر این سیستم‌ها را تمام‌گذر می‌نامند.  

مشخصات یک سیستم تمام‌گذر به طور کامل توسط جابجایی فاز آن تعیین می‌شود. البته در حالت کلی سیستم‌های LTI هم اندازه تبدیل فوریه را با بهره $|H(j\omega)|$ یا $|H(e^{j\omega})|$ تغییر می‌دهند و هم فاز آن را، و این تغییر فاز می‌تواند خطی یا غیر خطی باشد.


## ۶-۲-۲ تاخیر گروه

همان‌طور که در بخش ۶-۲-۱ گفتیم، سیستم‌هایی دارای تغییر فاز خطی این ویژگی را دارند که اثرشان را می‌توان با یک جابجایی زمانی ساده بیان کرد. در واقع طبق معادلات (۶-۸) و (۶-۹)، شیب فاز اندازه جابجایی زمانی را نشان می‌دهد.  

در حالت پیوسته در زمان، اگر  
$$
\angle H(j\omega) = -\omega t_0
$$  
آنگاه جابجایی به اندازه $-t_0$ است، یعنی تاخیری به اندازه $t_0$.  

در حالت گسسته در زمان:  
$$
\angle H(e^{j\omega}) = -\omega n_0
$$  
با تاخیر $n_0$ متناظر است.  

مفهوم تاخیر را می‌توان به طور طبیعی تعمیم داد تا فازهای غیر خطی را نیز شامل شود. فرض کنید بخواهیم اثر فاز یک سیستم LTI پیوسته در زمان را بر یک ورودی باریک‌باند بررسی کنیم. منظور از سیگنال باریک‌باند، سیگنالی است که تبدیل فوریه‌اش در خارج باند باریکی حول فرکانس $\omega = \omega_0$ صفر یا ناچیز است.  

اگر باند سیگنال خیلی باریک باشد می‌توانیم فاز سیستم را در این باند با تعریف خطی زیر بیان کنیم:  

$$
\angle H(j\omega) \approx -\varphi - \omega \alpha
$$  
(معادله ۶-۱۲)  

پس:  
$$
Y(j\omega) \approx X(j\omega) |H(j\omega)| e^{-j\varphi} e^{-j\omega \alpha}
$$  
(معادله ۶-۱۳)  

اثر تقریبی سیستم بر این سیگنال باریک‌باند، تغییر اندازه با ضریب $|H(j\omega)|$، ضرب در عامل مختلط ثابت، و ضرب در فاز خطی $e^{-j\omega \alpha}$ است که با تاخیر $\alpha$ ثانیه‌ای متناظر است.  

این تاخیر زمانی را **تاخیر گروه** در $\omega = \omega_0$ می‌نامند، زیرا تاخیر مشترکی را که فرکانس‌های باند حول فرکانس مرکزی $\omega = \omega_0$ پیدا می‌کنند، نشان می‌دهد.  

تاخیر گروه در هر فرکانسی با منفی شیب فاز در آن فرکانس برابر است، یعنی تاخیر گروه به صورت زیر تعریف می‌شود:  

$$
\tau(\omega) = -\frac{d}{d\omega} \{ \angle H(j\omega) \}
$$  
(معادله ۶-۱۴)  

مفهوم تاخیر گروه را می‌توان در مورد سیستم‌های گسسته در زمان نیز به کار برد. در مثال زیر اثر تاخیر گروه غیر ثابت بر یک سیگنال را در نظر می‌گیریم.

### مثال ۶-۱  

پاسخ ضربه یک سیستم تمام‌گذر با تاخیر گروه متغیر با فرکانس را در نظر بگیرید.  

پاسخ فرکانسی $H(j\omega)$ را حاصل‌ضرب سه عامل فرض می‌کنیم، یعنی:  

$$
H(j\omega) = \prod_{i=1}^{3} H_i(j\omega)
$$  

که در آن:  
(۶-۱۵) 
![معادله ۶-۱۵](https://raw.githubusercontent.com/tayebehemati/test/2f4a91ad23c18e92b314c9048883b0647a297ed9/images/%D9%81%D8%B1%D9%85%D9%88%D9%84%2015-6.jpg)  

غالباً بهتر است فرکانس $\omega_i$ با واحد رادیان بر ثانیه را بر حسب فرکانس $f_i$ با واحد هرتز بیان کنیم:  

$$
\omega_i = 2\pi f_i
$$  

در اینجا داریم:  
$$
f_1 \approx 50 \text{Hz}
$$  
$$
f_2 \approx 150 \text{Hz}
$$  
$$
f_3 \approx 300 \text{Hz}
$$  

چون صورت هر عامل $H_i(j\omega)$ مزدوج مختلط مخرج آن است، پس $|H_i(j\omega)| = 1$.  

پس می‌توانیم نتیجه بگیریم که فاز هر عامل $H_i(j\omega)$ را می‌توان از معادله (۶-۱۵) به دست آورد.  
و
![معادله ۱۵-۶](https://raw.githubusercontent.com/tayebehemati/test/66a8c2ef1a37e0d01a0aeedd63f9b1830bad9774/images/%D9%85%D8%B9%D8%A7%D8%AF%D9%84%D9%87%2015-6.jpg)  

اگر مقدار $\angle H(j\omega)$ را در فاصله $-\pi$ تا $\pi$ بگیریم، تابع فاز اصلی را مطابق شکل ۶-۵ (الف) به دست می‌آوریم؛ این شکل فاز را بر حسب فرکانس دارای واحد هرتز نشان می‌دهد.  

توجه کنید که این تابع در بعضی فرکانس‌ها ناپیوستگی‌هایی با اندازه $2\pi$ دارد، یعنی تابع فاز در این فرکانس‌ها مشتق‌پذیر نیست. ولی با افزودن مضارب صحیح $2\pi$ به فاز، پاسخ فرکانسی تغییر نمی‌کند.  

با افزودن مضارب صحیح مناسب $2\pi$ به بخش‌های مختلف فاز، فاز باز شده شکل ۶-۵ (ب) را به دست می‌آوریم. اکنون می‌توانیم تاخیر گروه را بر حسب فرکانس به صورت زیر حساب کنیم:  

$$
\tau(\omega) = -\frac{d}{d\omega} \left\{ \angle [H(j\omega)] \right\}
$$
 

که در آن \$\angle H(j\omega)\$ تابع باز شده فاز \$H(j\omega)\$ را نشان می‌دهد.

شکل ۶-۵ (ج) تاخیر گروه \$\tau(\omega)\$ را نشان می‌دهد. توجه کنید که فرکانس‌های حول \$50\text{Hz}\$ تاخیری بیشتر از فرکانس‌های حول \$150\text{Hz}\$ و \$300\text{Hz}\$ پیدا می‌کنند.

اثر این تاخیر گروه غیر ثابت را می‌توان به طور کیفی در پاسخ ضربه (شکل ۶-۵ (د)) سیستم LTI نیز مشاهده کرد.

می‌دانیم که:  
$$
F\{\delta(t)\} = 1
$$  
مولفه‌های فرکانسی ضربه چنان قرار می‌گیرند که ترکیبشان یک ضربه دست دهد و ضربه تابعی به شدت محلی است.  

چون سیستم تمام‌گذر تاخیر گروه غیر ثابت دارد، فرکانس‌های مختلف ورودی تاخیرهای متفاوتی می‌یابند.  

این پدیده را **پخش** می‌نامند.  

 در مثال فعلی بزرگ‌ترین تاخیر گروه در $50\ \text{Hz}$ رخ می‌دهد. پس بخش انتهایی پاسخ ضربه با فرکانسی حول $50\ \text{Hz}$ نوسان می‌کند.  

شکل ۶-۵ (و) به وضوح این پدیده را نشان می‌دهد.  

![شکل ۶-۵](https://raw.githubusercontent.com/tayebehemati/test/bc131e5c60c27dc8e74df01b62d83adc05ffaa91/images/%D8%B4%DA%A9%D9%84%206-5.jpg)


شکل ۶-۵: فاز، تاخیر گروه و پاسخ ضربه سیستم تمام‌گذر مثال ۶-۱:  
(الف) فاز اصلی  
(ب) فاز باز شده  
(ج) تاخیر گروه  
(د) پاسخ ضربه  
تمام این کمیت‌ها بر حسب فرکانس هرتزی رسم شده‌اند.



### مثال ۶-۲  

تاخیر گروه غیر ثابت از عامل‌هایی است که در شبکه‌های مخابراتی قطع و وصلی با اهمیت شمرده می‌شود.  
شرکت AT&T/Bell نتایج بررسی خود در تمام فلات آمریکا راجع به تاخیر گروه مخابره‌های مختلف گزارش کرده است.  

شکل ۶-۶ برخی نتایج این بررسی را برای دو دسته مخابره این‌گونه نشان می‌دهد.  
آنچه در منحنی‌های شکل ۶-۶ (الف) نشان داده شده، در واقع بخش غیر ثابت تاخیر گروه است.  
یعنی تاخیر ثابتی برابر با مینیمم تاخیر در فرکانس‌های مختلف از تاخیر گروه کم شده و تفاضل حاصل در شکل ۶-۶ (الف) نشان داده شده است.  

بنابراین هر منحنی شکل ۶-۶ (الف)، تاخیر اضافی مؤلفه‌های فرکانسی مختلف هر دسته را نشان می‌دهد.  
منحنی‌ها با کلمه‌های short و medium متمایز شده‌اند که به ترتیب مخابرات راه نزدیک (۰ تا ۱۸۰ مایل هوایی) و راه متوسط (۱۸۰ تا ۷۲۵ مایل هوایی) را نشان می‌دهد.  

می‌بینیم که تاخیر گروه در فرکانس $1700\ \text{Hz}$ مینیمم است و در دو طرف این فرکانس به طور یکنوا افزایش می‌یابد.  

![شکل ۶-۶](https://raw.githubusercontent.com/tayebehemati/test/1046a7337f66aca5392851c1e24db63b47678106/images/%D8%B4%DA%A9%D9%84%206-6.jpg)

**شکل ۶-۶** (الف) بخش غیر ثابت تاخیر گروه؛ و (ب) اندازه پاسخ فرکانسی تلفن‌های راه نزدیک و راه متوسط (گرفته شده از مقاله Duffy و Thatcher).  
این منحنی‌ها بر حسب فرکانس رسم شده‌اند و فرکانس با واحد هرتز مشخص شده است.  
دامنه پاسخ فرکانسی نیز مطابق معمول بر حسب دسیبل است. یعنی $20 \log_{10} \left( | H(j\omega) | \right)$ رسم شده است.  
این گونه منحنی‌ها را در بخش ۶-۲-۳ به تفصیل بررسی می‌کنیم.  

---

با ترکیب مشخصات تاخیر گروه شکل ۶-۶ (الف) و پاسخ فرکانسی شکل ۶-۶ (ب)، پاسخ ضربه‌های شکل ۶-۷ را به دست می‌آوریم.  
پاسخ ضربه شکل ۶-۷ (الف) به دسته راه نزدیک تعلق دارد.  
مؤلفه‌های بسیار کم‌فرکانس و بسیار پُرفرکانس زودتر از مؤلفه‌های فرکانس متوسط بروز می‌کنند.  
این ویژگی با آنچه شکل ۶-۶ (الف) نشان می‌دهد تطابق دارد.  
شکل ۶-۷ (ب) همین پدیده را برای تلفن‌های راه متوسط نشان می‌دهد.  

![شکل ۶-۷](https://raw.githubusercontent.com/tayebehemati/test/c482dd463c2997543e982a960d9d5db8ed7fe0be/images/%D8%B4%DA%A9%D9%846-7.jpg)

**شکل ۶-۷** پاسخ ضربه متناظر با تاخیر گروه و پاسخ فرکانسی نشان داده شده در شکل ۶-۶:  
(الف) پاسخ ضربه متناظر با تلفن‌های راه نزدیک؛  
(ب) پاسخ ضربه متناظر با تلفن‌های راه دور.
