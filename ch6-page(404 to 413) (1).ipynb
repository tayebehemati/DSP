</div>

<div dir="rtl" style="text-align: right;">

از شکل ۶-۱۸ (الف) می‌توان نتیجه گرفت که باند گذار فیلتر باترورث عریض‌تر از فیلتر بیضوی است.
بهایی که در ازای باریکی باند گذار در فیلتر بیضوی پرداخت شده، در شکل ۶-۱۸ (ب) مشخص است؛ این شکل پاسخ پله دو فیلتر را نشان می‌دهد. دیده می‌شود که نوسان‌های میرای پاسخ پله فیلتر بیضوی شدیدتر از فیلتر باترورث است. به‌ویژه، زمان نشست پاسخ پله در فیلتر بیضوی طولانی‌تر است.
</div>

# ۵-۶ سیستم‌های پیوسته در زمان مرتبه اول و دوم
<div dir="rtl" style="text-align: right;">
سیستم‌های LTI که با معادلات دیفرانسیل خطی با ضرایب ثابت توصیف می‌شوند، کاربرد فراوانی دارند؛ زیرا بسیاری از سیستم‌های فیزیکی را می‌توان با این معادلات مدل‌سازی کرد و پیاده‌سازی آن‌ها نیز ساده است.
به دلایل عملی، سیستم‌های مرتبه بالاتر معمولاً به‌صورت ترکیب‌های سری یا موازی از سیستم‌های مرتبه اول و دوم ساخته می‌شوند. بنابراین، شناخت دقیق سیستم‌های مرتبه اول و دوم برای تحلیل و طراحی سیستم‌های پیچیده‌تر ضروری است.
در این بخش، به بررسی سیستم‌های مرتبه پایین می‌پردازیم. همچنین در بخش ۶-۶، معادل‌های گسسته‌زمان این سیستم‌ها معرفی خواهند شد.

#### ۶-۵-۶ سیستم‌های درجه اول

<div dir="rtl" style="text-align: right;">
معادله دیفرانسیل سیستم درجه اول معمولاً به صورت زیر نوشته می‌شود:
</div>

$$
r \frac{dy(t)}{dt} + y(t) = x(t) \tag{۶-۲۱}
$$
<div dir="rtl" style="text-align: right;">
در این رابطه، $r$ ضریبی است که اهمیت آن بعداً مشخص خواهد شد.
پاسخ فرکانسی سیستم درجه اول به صورت زیر است:
</div>

$$
H(j\omega) = \frac{۱}{j\omega\tau + ۱} \tag{۶-۲۱}
$$

<div style="text-align: center;">
    <img src="https://raw.githubusercontent.com/sepidehrezvani/page-404-408-/75e861ca4987ace9cf673b08e2542620510bc2bf/%D8%B4%DA%A9%D9%84%206-18%20%D8%A7%D9%84%D9%81%20%D8%B5%D9%81%D8%AD%D9%87%20405.png"
         alt="شکل ۶-۲۱"
         style="display: block; margin: auto; width: 50%;"> 
<div style="text-align: center;">
    <img src="https://raw.githubusercontent.com/sepidehrezvani/page-404-408-/75e861ca4987ace9cf673b08e2542620510bc2bf/%D8%B4%DA%A9%D9%84%206-18%20%D8%A8%20%D8%B5%D9%81%D8%AD%D9%87%20405.png"
         alt="شکل ۶-۲۱"
         style="display: block; margin: auto; width: 50%;"> 
<p style="font-size:10px; text-align:center;">
شکل ۶-۱۸ نمونه هایی از فیلتر با ترورث و بیضوی مرتبه پنجم که تموج باند عبور ، تموج باند حذف ، و فرکانس قطع  یکسانی دارند: (الف) اندازه های پاسخ فرکانسی بر حسب فرکانس فرکانس بر حسب هرتز ؛ (ب) پاسخ پله.  
</p>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, cheby1, ellip, freqz

# پارامترهای فیلتر
order = 5
fs = 4000  # نرخ نمونه‌برداری (برای تعیین مقیاس فرکانس)
nyq = fs / 2
cutoff = 600  # فرکانس قطع (Hz)
wn = cutoff / nyq  # نرمال‌شده

# طراحی فیلترها
b_butter, a_butter = butter(order, wn, btype='low', analog=False)
b_cheby1, a_cheby1 = cheby1(order, 0.5, wn, btype='low', analog=False)
b_ellip, a_ellip = ellip(order, 0.5, 40, wn, btype='low', analog=False)

# پاسخ فرکانسی
w, h_butter = freqz(b_butter, a_butter, worN=1024, fs=fs)
_, h_cheby1 = freqz(b_cheby1, a_cheby1, worN=1024, fs=fs)
_, h_ellip = freqz(b_ellip, a_ellip, worN=1024, fs=fs)

# رسم نمودار
plt.figure(figsize=(10, 6))
plt.plot(w, np.abs(h_cheby1), 'k-', label='فیلتر چبیشف نوع اول')
plt.plot(w, np.abs(h_ellip), 'k', linestyle='-', linewidth=1, label='فیلتر بیضوی')
plt.plot(w, np.abs(h_butter), 'k--', linewidth=1.5, label='فیلتر باترورث')

# تنظیمات نمودار
plt.xlabel('فرکانس (Hz)')
plt.ylabel('اندازه پاسخ فرکانسی')
plt.ylim(0, 1.05)
plt.xlim(0, 2000)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='best', fontsize=11)
plt.tight_layout()
plt.show()

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, bessel, step

# تنظیمات زمان
t = np.linspace(0, 0.02, 1000)  # 0 تا 20 میلی‌ثانیه
fs = 50000  # فرکانس نمونه‌برداری
order = 5  # مرتبه فیلتر
fc = 1000  # فرکانس قطع (Hz)

# طراحی فیلتر Butterworth و Bessel
b_butter, a_butter = butter(order, 2 * np.pi * fc, analog=True)
b_bessel, a_bessel = bessel(order, 2 * np.pi * fc, analog=True)

# پاسخ پله هر دو فیلتر
t1, y_butter = step((b_butter, a_butter), T=t)
t2, y_bessel = step((b_bessel, a_bessel), T=t)

# رسم نمودار
plt.figure(figsize=(10, 6))
plt.plot(t1 * 1000, y_butter, 'k-', linewidth=2, label='فیلتر باتروث')  # خط پیوسته
plt.plot(t2 * 1000, y_bessel, 'k--', linewidth=2, label='فیلتر بسل')    # خط‌چین

# برچسب و تنظیمات
plt.xlabel('(msec) زمان', fontsize=13)
plt.ylabel('', fontsize=13)
plt.title('', fontsize=14)
plt.xlim(0, 20)
plt.ylim(0, 1.2)
plt.legend(loc='lower right', fontsize=12)
plt.grid(True)

plt.show()
```

<div dir="rtl" style="text-align: right;">
و پاسخ ضربه آن عبارت است از:
</div>

$$
h(t) = \frac{۱}{\tau} e^{-\frac{t}{\tau}} u(t) \tag{۶-۲۳}
$$
<div dir="rtl" style="text-align: right;">
که در شکل ۶-۱۹ (الف) رسم شده است. 
پاسخ پله سیستم عبارت است از
</div>

$$
s(t) = h(t) * u(t) = \left[۱ - e^{-\frac{t}{\tau}} \right] u(t) \tag{۶-۲۴}
$$

<div dir="rtl" style="text-align: right;">
که در شکل ۴-۳۹ (ب) رسم شده است. پارامتر $\tau$ <b>ثابت زمانی</b> سیستم است و آهنگ پاسخ سیستم مرتبه اول را کنترل می‌کند. 
برای مثال، چنانچه شکل ۶-۱۹ نشان می‌دهد، در $t = \tau$، پاسخ ضربه به $\frac{۱}{e}$ اندازه پاسخ در $t = ۰$ و پاسخ پله به $\frac{۱}{e}$ مقدار نهایی‌اش می‌رسد. بنابراین هرچه $\tau$ کوچکتر باشد، پاسخ ضربه سریع‌تر نزول می‌کند و زمان صعود پاسخ پله کمتر می‌شود؛ یعنی پاسخ پله سریع‌تر به مقدار نهایی خود می‌رسد. 
همچنین توجه کنید که پاسخ پله سیستم مرتبه اول نوسان میرا ندارد.
</div>
<div dir="rtl" style="text-align: right;">
شکل ۶-۲۰ نمودار بودی پاسخ فرکانسی معادله زیر را نشان می‌دهد
در این شکل، یک مزیت به کار بردن مقیاس لگاریتمی برای فرکانس نشان داده شده است. این امکان را به ما می‌دهد که بدون زحمت زیاد تقریب مفیدی از نمودار سیستم مرتبه اول بدست آوریم. برای پی بردن به این مطلب ابتدا نمودار لگاریتم دامنه پاسخ فرکانسی را بررسی می‌کنیم. از معادله بالا به دست می‌آوریم:
</div>

$$
۲۰ \log_{۱۰} |H(jω)| = -۱۰ \log_{۱۰} \left[ (\omega\tau)^۲ + ۱ \right] \tag{۶-۲۵}
$$
<div dir="rtl" style="text-align: right;">
می‌بینیم که به‌ازای $\omega\tau \ll 1$، لگاریتم دامنه تقریباً صفر است، در حالی که به‌ازای $\omega\tau \gg ۱$، لگاریتم دامنه تقریباً تابعی خطی از $\log_{۱۰}(\omega)$ است. یعنی
</div>
<div dir="rtl" style="text-align: right;">
به ازای 
</div>

$$  ۲۰ \log_{۱۰} |H(j\omega)| \approx ۰   \quad  \omega \ll \frac{1}{\tau} \tag{۶-۲۶}  $$
<div dir="rtl" style="text-align: right;">
و به ازای
</div>

$$  ۲۰ \log_{۱۰}|H(j\omega)| \approx -۲۰ \log_{۱۰}(\omega\tau) = -۲۰ \log_{۱۰}(\omega) -۲۰ \log_{۱۰}(\tau) \quad \omega \gg \frac{۱}{\tau} \tag{۶-۲} $$
</div>

         style="display: block; margin: auto; width: 50%;"> 
<div style="text-align: center;">

         style="display: block; margin: auto; width: 50%;"> 
<p style="font-size:10px; text-align:center;">
.شکل ۶-۱۹  سیستم مرتبه اول در زمان: (الف)پاسخ ضربه؛ (ب)پاسخ پله
</p>

```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات
tau = 1
t = np.linspace(0, 5*tau, 500)

# (الف) پاسخ ضربه h(t) = (1/τ) * exp(-t/τ)
h = (1/tau) * np.exp(-t/tau)

# (ب) پاسخ پله s(t) = 1 - exp(-t/τ)
s = 1 - np.exp(-t/tau)

# رسم نمودار
fig, axs = plt.subplots(2, 1, figsize=(7, 8))

# نمودار (الف)
axs[0].plot(t, h, 'k', linewidth=2)
axs[0].hlines(1/tau, 0, 0.1, colors='k')
axs[0].hlines(1/(tau*np.e), 0, tau, colors='k', linestyles='dashed')
axs[0].axvline(tau, color='k', linestyle='dashed')
axs[0].text(0.02, 1/tau + 0.05, r'$1/\tau$', fontsize=12)
axs[0].text(tau + 0.1, 1/(tau*np.e), r'$1/(\tau e)$', fontsize=12)
axs[0].text(tau - 0.05, -0.02, r'$\tau$', fontsize=12)
axs[0].set_ylabel('h(t)', fontsize=14)
axs[0].set_xticks([])
axs[0].set_yticks([])
axs[0].set_xlim(0, 5*tau)
axs[0].set_ylim(0, 1.2/tau)
axs[0].set_title('(الف)', loc='right', fontsize=14)

# نمودار (ب)
axs[1].plot(t, s, 'k', linewidth=2)
axs[1].hlines(1, 0, 0.1, colors='k', linestyles='dashed')
axs[1].hlines(1 - 1/np.e, 0, tau, colors='k', linestyles='dashed')
axs[1].axvline(tau, color='k', linestyle='dashed')
axs[1].text(0.05, 1.02, '1', fontsize=12)
axs[1].text(tau + 0.1, 1 - 1/np.e, r'$1 - \frac{1}{e}$', fontsize=12)
axs[1].text(tau - 0.05, -0.05, r'$\tau$', fontsize=12)
axs[1].set_ylabel('s(t)', fontsize=14)
axs[1].set_xlabel('t', fontsize=14)
axs[1].set_xticks([])
axs[1].set_yticks([])
axs[1].set_xlim(0, 5*tau)
axs[1].set_ylim(0, 1.1)
axs[1].set_title('(ب)', loc='right', fontsize=14)

plt.tight_layout()
plt.show()
```

<div dir="rtl" style="text-align: right;">
به بیان دیگر برای سیستم درجه اول مجانیهای لگاریتم دامنه در فرکانسهای پایین و بالا، خط راست اند.  
مجانب فرکانس پایین [معادله (۶-۲۶)] خط $ ۰dB$  
و مجانب فرکانس بالا [معادله (۶-۲۷)] خطی است که به ازای هر دهه افزایش (یعنی ۱۰ برابر شدن آن) $  |H(j\omega)|  $ به اندازه $ ۲۰dB$ افت می‌کنند.  
این مجانب را گاهی مجانب $ ۲۰dB $ بر دهه می‌نامند.  
</div>
<div dir="rtl" style="text-align: right;">
توجه کنید که دو تقریب مجانبی بیان شده با معادلات (۶-۲۶) و (۶-۲۷) در نقطه  
$ \log_{۱۰}(j\omega) = -\log_{۱۰}(\tau) $ با $ \omega = \frac{۱}{\tau} $ مساوی اند.  
از لحاظ ترسیمی دو مجانب همدیگر را در نقطه $ \omega = \frac{۱}{\tau} $ قطع می‌کنند،  
تقریب مجانبی لگاریتم دامنه با دو خط راست با توجه به این مطلب به دست می‌آید، تقریب  
$ ۲۰ \log_{۱۰} |H(j\omega)| $ با صفر در $ \omega \ll \frac{۱}{\tau} $ و با معادله (۶-۲۷) در $ \omega \gg \frac{۱}{\tau} $.  
این تقریب نیز در شکل ۶-۲۰ (با خط چین) نشان داده شده است.  
نقطه‌ای که در آن شیب تقریب تغییر می‌کند، دقیقاً $ \omega = \frac{۱}{\tau} $ است که به همین دلیل آن را فرکانس شکست می‌نامند.  
همچنین توجه کنید که در $ \omega = \frac{۱}{\tau} $ دو جمله آرگومان لگاریتم معادله (۶-۲۵) برابرند.
</div>
<div dir="rtl" style="text-align: right;">
پس در این نقطه  
مقدار واقعی اندازه عبارت است از
</div>

$$ ۲۰ \log_{۱۰} |H(j \frac{۱}{\tau})| = -۱۰ \log_{۱۰}(۲) \approx -۳ \, \text{dB} $$

```python
import numpy as np
import matplotlib.pyplot as plt

# محدوده فرکانس
w = np.logspace(-1, 2, 500)  # از 0.1 تا 100 (برحسب 1/τ)

# تابع انتقال فیلتر پایین‌گذر درجه اول
H = 1 / np.sqrt(1 + w**2)

# پاسخ دسی‌بل
H_db = 20 * np.log10(H)

# تقریب مایانی: برای w > 1/τ => H ~ 1/w
w_approx = w[w >= 1]
H_approx_db = -20 * np.log10(w_approx)

# رسم
plt.figure(figsize=(10, 5))
plt.semilogx(w, H_db, 'k', linewidth=2, label=r'$|H(j\omega)|$')
plt.semilogx(w_approx, H_approx_db, 'k--', linewidth=1.5, label='تقریب میانی')

# خطوط راهنما برای 3dB
plt.plot([1, 1], [-3, 0], 'k:', linewidth=1)
plt.plot([0.9, 1.1], [-3, -3], 'k:', linewidth=1)
plt.text(1.05, -2, '3 dB', fontsize=12)

# تنظیمات محور و برچسب‌ها
plt.ylim(-60, 10)
plt.xlim(0.1, 100)
plt.xticks([0.1, 1, 10, 100], [r'$0.1/\tau$', r'$1/\tau$', r'$10/\tau$', r'$100/\tau$'])
plt.yticks(np.arange(-60, 21, 20))
plt.xlabel(r'$\omega$', fontsize=14)
plt.ylabel(r'$20 \log_{10} |H(j\omega)|$', fontsize=14)
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.legend(loc='upper right', fontsize=11)
plt.tight_layout()
plt.show()

import numpy as np
import matplotlib.pyplot as plt

# پارامتر زمانی
tau = 1
omega = np.logspace(-1, 2, 500)  # از 0.1 تا 100 (در مقیاس لگاریتمی)

# تابع تبدیل: H(jω) = 1 / (1 + jωτ)
H = 1 / (1 + 1j * omega * tau)
phase = np.angle(H)  # فاز واقعی

# تقریب مداری (فاز تقریبی)
phase_approx = np.piecewise(
    omega,
    [omega < 1/tau, (omega >= 1/tau) & (omega <= 10/tau), omega > 10/tau],
    [0, lambda w: -np.pi/4 * np.log10(w * tau * 10), -np.pi/2]
)

# رسم نمودار
plt.figure(figsize=(9, 5))
plt.semilogx(omega/tau, phase, 'k', label='فاز دقیق', linewidth=2)  # نمودار اصلی
plt.semilogx(omega/tau, phase_approx, 'k--', label='تقریب مداری', linewidth=1.5)  # نمودار خط‌چین

# برچسب محور و مقیاس
plt.yticks(
    [-3*np.pi/4, -np.pi/2, -np.pi/4, 0, np.pi/4],
    [r'$-\frac{3\pi}{4}$', r'$-\frac{\pi}{2}$', r'$-\frac{\pi}{4}$', '0', r'$\frac{\pi}{4}$']
)
plt.xticks([0.1, 1, 10, 100], [r'$0.1/\tau$', r'$1/\tau$', r'$10/\tau$', r'$100/\tau$'])
plt.xlabel(r'$\omega$', fontsize=14)
plt.ylabel(r'$\angle H(j\omega)$', fontsize=14)

# متن فارسی
plt.text(0.15, -0.15, 'تقریب مجانی', fontsize=12)

plt.grid(True, which='both', linestyle='--', alpha=0.3)
plt.legend(loc='lower left')
plt.ylim(-3*np.pi/4, np.pi/4)
plt.xlim(0.08, 120)
plt.tight_layout()
plt.show()
```

<div dir="rtl" style="text-align: right;">
به همین خاطر گاهی نقطه $ \omega = \frac{۱}{\tau} $ را نقطه $۳dB$ می‌نامند. شکل نشان می‌دهد که نمودار بودی تنها در حوالی فرکانس شکست با مجانب خط راست تفاوت زیادی دارد.  
بنابراین برای ترسیم دقیق‌تر نمودار بودی تنها باید مجانب‌ها را نزدیک فرکانس شکست اصلاح کنیم.  
</div>

<div dir="rtl" style="text-align: right;">
برای $ \angle H(j\omega) $ نیز می‌توان تقریب خط راست مفیدی به دست آورد:  
</div>

$$
\angle H(j\omega) = -\tan^{-۱} (\omega\tau) \approx
\begin{cases}
0 & \omega \leq \frac{۰/۱}{\tau} \\
-\frac{\pi}{۴} [\log_{۱۰}(\omega\tau) + ۱] & \frac{۰/۱}{\tau} \leq \omega \leq \frac{۱۰}{\tau} \\
-\frac{\pi}{۲} & \omega \geq \frac{۱۰}{\tau}
\end{cases}
$$
<div dir="rtl" style="text-align: right;">
توجه کنید که در فاصله
</div>
    
$$ \frac{۰/۱}{\tau} \leq \omega \leq \frac{۱۰}{\tau} $$ 
<div dir="rtl" style="text-align: right;">
مقدار این تقریب (از صفر به $ -\frac{\pi}{۲} $) به صورت خطی کاهش می‌یابد.  
مجانب واقعی $ \angle H(j\omega) $ که به ازای $ \omega \gg \frac{۱}{\tau} $ صفر و مجانب واقعی $ \angle H(j\omega) $ به ازای $ \omega \ll \frac{۱}{\tau} $، $ -\frac{\pi}{۲} $ است.  
به علاوه تقریب در فرکانس شکست $ \omega = \frac{۱}{\tau} $ با مقدار واقعی برابر است.  
در این نقطه داریم  
</div>
 
$$ \angle H(j \frac{۱}{\tau}) = -\frac{\pi}{۴} \tag{۱۶۸-۴} $$
<div dir="rtl" style="text-align: right;">
تقریب مجانبی فاز نیز در شکل ۶-۲۰ رسم شده است و این شکل چگونگی اصلاح مجانب‌های راست برای دقیق‌تر شدن منحنی $ \angle H(j\omega) $ را نیز نشان می‌دهد.  
</div>

<div dir="rtl" style="text-align: right;">
با توجه به این سیستم مرتبه اول، بار دیگر می‌توان رابطه معکوس بین حوزه‌های زمان و فرکانس را مشاهده کرد.  
هر چه $ \tau $ را کوچک‌تر کنیم، پاسخ زمانی سیستم سریع‌تر می‌شود  
(یعنی $ h(t) $ به طرف مبدأ فشرده‌تر و زمان صعود پاسخ پله کوچک‌تر می‌شود)  
و به‌طور هم‌زمان فرکانس شکست بزرگ‌تر می‌شود  
(یعنی $ H(j\omega) $ بهتر می‌شود زیرا روی گستره فرکانس بزرگ‌تری $ |H(j\omega)| \approx 1 $ باقی می‌ماند).  
با ضرب پاسخ ضربه در $ \tau $ و مقایسه $ \tau h(t) $ و $ H(j\omega) $ هم می‌توان این نکته را دریافت:  
</div>

$$
\tau h(t) = e^{-t/\tau} \, u(t)
\quad \text{و} \quad
H(j\omega) = \frac{۱}{j\omega\tau + ۱}
$$
<div dir="rtl" style="text-align: right;">
به این ترتیب $\tau h(t)$ تابعی از   $t/\tau$ و $H(j\omega)$  تابعی از  $\omega\tau$ می‌شود. بنابراین تغییر $\tau$ با تغییر مقیاس  زمان و فرکانس متناظرست. 
</div>

<div dir="rtl" style="text-align: right;">

### ۲-۵-۶ سیستم درجه دوم  
<div dir="rtl" style="text-align: right;">
معادله دیفرانسیل خطی با ضرایب ثابت سیستم درجه دوم عبارت است از  
</div>

  $$
\frac{d^۲ y(t)}{dt^2} + ۲ζω_n \frac{dy(t)}{dt} + ω_n^۲ y(t) = ω_n^۲ x(t) \tag{۶-۳۱}
$$


<div dir="rtl" style="text-align: right;">
بسیاری از سیستم‌های فیزیکی، مثلاً مدارهای RLC و سیستم‌های مکانیکی نظیر شکل ۶-۲۱، متشکل از فنر، جرم و میراساز چسبنده یا ضربه‌گیر، چنین معادلاتی دارند. در این شکل، ورودی نیروی اعمال‌شده $x(t)$ و خروجی $y(t)$ جابه‌جایی جرم از نقطه تعادلی است که در آن هیچ نیرویی به فنر وارد نمی‌شود. معادله حرکت این سیستم چنین است:  
</div>

$$
m \frac{d^۲ y(t)}{dt^۲} = x(t) - ky(t) - b \frac{dy(t)}{dt}
$$
<div dir="rtl" style="text-align: right;">
یا
</div>

$$
\frac{d^۲ y(t)}{dt^۲} + \left( \frac{b}{m} \right) \frac{dy(t)}{dt} + \left( \frac{k}{m} \right) y(t) = \frac{۱}{m} x(t)
$$

<div dir="rtl" style="text-align: right;">
مقایسه این معادله و معادله (۶-۳۱) مقادیر زیر را به دست می‌دهد:
</div>

$$
ω_n = \sqrt{\frac{k}{m}} \tag{۶-۳۲}
$$

<div dir="rtl" style="text-align: right;">
و
</div>

$$
ζ = \frac{b}{۲\sqrt{km}}
$$

<div dir="rtl" style="text-align: right;">
پس معادله حرکت سیستم شکل ۶-۲۱ با معادله (۶-۳۱) فقط در ضریب مقیاس $x(t)$ تفاوت دارد. پاسخ فرکانسی سیستم مرتبه دوم معادله (۶-۳۱) عبارت است از:
</div>

$$
H(jω) = \frac{ω_n^۲}{(jω)^۲ + ۲ζω_n (jω) + ω_n^۲} \tag{۶-۳۳}
$$

<div dir="rtl" style="text-align: right;">
مخرج را می‌توان به صورت زیر تجزیه کرد:
</div>

$$
H(jω) = \frac{ω_n^۲}{(jω - c_۱)(jω - c_۲)}
$$

<div dir="rtl" style="text-align: right;">
که در آن:
</div>

$$
c_۱ = -ζω_n + ω_n \sqrt{ζ^۲ - ۱} \tag{۶-۳۴}
$$
$$
c_۲ = -ζω_n - ω_n \sqrt{ζ^۲ - ۱}
$$
<div dir="rtl" style="text-align: right;">
به ازای $ζ ≠ ۱$، $c_۱$ و $c_۲$ مختلف‌اند و می‌توانیم $H(jω)$ را به صورت کسرهای جزئی زیر بسط دهیم:
</div>

$$
H(jω) = \frac{M}{jω - c_۱} - \frac{M}{jω - c_۲} \tag{۶-۳۵}
$$
<div dir="rtl" style="text-align: right;">
که در آن:
</div>

$$
M = \frac{ω_n}{۲\sqrt{ζ^۲ - ۱}} \tag{۶-۳۶}
$$
<div dir="rtl" style="text-align: right;">
پاسخ ضربه سیستم را می‌توان به‌سادگی از معادله (۶-۳۵) به دست آورد:
</div>

$$
h(t) = M \left[e^{c_۱ t} - e^{c_۲ t} \right] u(t) \tag{۶-۳۷}
$$

<div dir="rtl" style="text-align: right;">
به ازای $ζ = ۱$، داریم: $c_۱ = c_۲ = -ω_n$
</div>

$$
H(jω) = \frac{ω_n^۲}{(jω + ω_n)^۲} \tag{۶-۳۸}
$$
<div dir="rtl" style="text-align: right;">
جدول ۴-۲ نشان می‌دهد که پاسخ ضربه در این حالت عبارت است از:
</div>

$$
h(t) = ω_n^۲ t e^{-ω_n t} u(t) \tag{۶-۳۹}
$$

<div dir="rtl" style="text-align: right;">
توجه کنید که معادلات (۶-۳۷) و (۶-۳۹) نشان می‌دهند که ${h(t)/ω_n}$ تابعی از $ω_n t$ است.  
به علاوه، معادله (۶-۳۳) را می‌توان به صورت زیر بازنویسی کرد:
</div>

$$
H(jω) = \frac{۱}{(jω / ω_n)^۲ + ۲ζ(jω / ω_n) + ۱}
$$
<div dir="rtl" style="text-align: right;">   
بنابراین پاسخ فرکانسی تابعی از ${ω /ω_n}$ است. تغییر $\omega_n$ معادل تغییر مقیاس زمانی و فرکانسی است.
</div>

</div>


<div dir="rtl" style="text-align: right;">
پارامتر $ζ$ را نسبت میرائی و پارامتر $ω_n$ را فرکانس طبیعی نامیرا می‌نامند. علت این نام‌گذاری با بررسی دقیق‌تر پاسخ فرکانسی و پاسخ پله سیستم مرتبه دوم روشن می‌شود. اول این که طبق معادله (۶-۳۵) می‌بینیم که به‌ازای $۰< ζ <۱$، $c_۱$ و $c_۲$ مختلط‌اند و می‌توانیم پاسخ ضربه را از معادله (۶-۳۷) به‌صورت زیر بنویسیم.
</div>

$$
h(t)=\frac{ω_n e^{-ζω_n t}}{۲j\sqrt{۱-ζ^۲}} \left\{ \exp\left[j(ω_n \sqrt{۱-ζ^۲})t\right] - \exp\left[-j(ω_n \sqrt{۱-ζ^۲})t\right] \right\} u(t)
\tag{۶-۴۰}
$$

$$
= \frac{ω_n e^{-ζω_n t}}{\sqrt{۱-ζ^۲}} \sin(ω_n \sqrt{۱-ζ^۲} \; t) \; u(t)
$$


<div dir="rtl" style="text-align: right;">
پس به‌ازای $۰<ζ<۱$ پاسخ ضربه سیستم مرتبه دوم نوسان میرا دارد و در این حالت سیستم را «زیر میرا» می‌نامند. به‌ازای $ζ>۱$، $c_۱$ و $c_۲$ حقیقی‌اند و پاسخ ضربه سیستم اختلاف دو تابع نمایی میراست. در این حالت سیستم را «فوق میرا» می‌نامند. به‌ازای $ζ=۱$، $c_۱$ و $c_۲$ برابرند و سیستم را «میرای بحرانی» می‌نامند.
شکل ۶-۲۲ (الف) پاسخ ضربه سیستم مرتبه دوم ضرب شده در $۱/ω_n$ را به‌ازای مقادیر مختلف $ζ$ نشان می‌دهد.
</div>
<div dir="rtl" style="text-align: right;">
پاسخ پله سیستم مرتبه دوم را می‌توان به‌ازای $ζ≠۱$ از معادله (۶-۳۷) به‌دست آورد. نتیجه عبارت است از:
</div>

$$
s(t) = h(t) * u(t) = \left\{ ۱ + M \left[ \frac{e^{c_۱ t}}{c_۱} - \frac{e^{c_۲ t}}{c_۲} \right] \right\} u(t)
\tag{۶-۴۱}
$$
<div dir="rtl" style="text-align: right;">
به‌ازای $ζ=۱$ می‌توانیم از معادله (۶-۳۹) استفاده کنیم:
</div>

$$
s(t) = \left[ ۱ - e^{-ω_n t} - ω_n t e^{-ω_n t} \right] u(t)
\tag{۶-۴۲}
$$

<div style="text-align: center;">
    <img src="https://raw.githubusercontent.com/tayebehemati/DSP/d93be13f4ab53bcb25d8f340d303ccbc445c2ddb/images/%D8%B4%DA%A9%D9%846-22.jpg"
         alt="شکل ۶-۲۲"
         style="display: block; margin: auto; width: 50%;">  
<p style="font-size:10px; text-align:center;">
شکل ۶-۲۲  پاسخ  سیستمهای مرتبه دوم پیوسته  در زمان به ازای مقادیر  مختلف نسبت میرایی :  (الف) پاسخ ضربه ؛ (ب) پاسخ پله  
</p>

```python
# تعریف تابع پاسخ ضربه h(t) برای سیستم مرتبه دوم
def impulse_response(t, zeta, omega_n):
    if zeta < 1:
        omega_d = omega_n * np.sqrt(1 - zeta**2)
        return (omega_n / np.sqrt(1 - zeta**2)) * np.exp(-zeta * omega_n * t) * np.sin(omega_d * t)
    elif zeta == 1:
        return omega_n**2 * t * np.exp(-omega_n * t)
    else:
        s1 = -omega_n * (zeta - np.sqrt(zeta**2 - 1))
        s2 = -omega_n * (zeta + np.sqrt(zeta**2 - 1))
        return omega_n**2 * (np.exp(s1 * t) - np.exp(s2 * t)) / (s2 - s1)

# رسم پاسخ ضربه
plt.figure(figsize=(10, 6))
for zeta in zeta_values:
    h = impulse_response(t, zeta, omega_n)
    plt.plot(t, h / omega_n, label=f'ζ = {zeta}')  # نرمال‌سازی با omega_n طبق نمودار اصلی

# تنظیمات ظاهری نمودار
plt.xlabel('t')
plt.ylabel(r'$h(t)/\omega_n$')
plt.xticks([1, 2], [r'$\frac{1}{\omega_n}$', r'$\frac{2}{\omega_n}$'])
plt.title('(الف)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

import numpy as np
import matplotlib.pyplot as plt

# زمان نرمال‌شده
t = np.linspace(0, 10, 1000)

# فرکانس طبیعی
omega_n = 1

# ضریب میرایی‌های مختلف
zeta_values = [0.1, 0.2, 0.4, 0.7, 1.0, 1.5]

# تابع پاسخ گذرا s(t)
def step_response(t, zeta, omega_n):
    if zeta < 1:
        omega_d = omega_n * np.sqrt(1 - zeta**2)
        return 1 - (1 / np.sqrt(1 - zeta**2)) * np.exp(-zeta * omega_n * t) * np.sin(omega_d * t + np.arccos(zeta))
    elif zeta == 1:
        return 1 - (1 + omega_n * t) * np.exp(-omega_n * t)
    else:
        s1 = -omega_n * (zeta - np.sqrt(zeta**2 - 1))
        s2 = -omega_n * (zeta + np.sqrt(zeta**2 - 1))
        A = s2 / (s2 - s1)
        B = -s1 / (s2 - s1)
        return 1 - A * np.exp(s1 * t) - B * np.exp(s2 * t)

# رسم نمودار
plt.figure(figsize=(10, 6))
for zeta in zeta_values:
    s = step_response(t, zeta, omega_n)
    plt.plot(t, s, label=f'ζ = {zeta}')

# تنظیمات ظاهری نمودار
plt.xlabel('t')
plt.ylabel('s(t)')
plt.xticks([1, 2, 3], [r'$\frac{1}{\omega_n}$', r'$\frac{2}{\omega_n}$', r'$\frac{3}{\omega_n}$'])
plt.title('(ب)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```

<div dir="rtl" style="text-align: right;">
توجه کنید که، چنان‌چه قبلاً گفتیم، مقدار $ω_n$ مقیاس زمانی $h(t)$ و $s(t)$ را کنترل می‌کند. مثلاً در حالت زیر میرا هر چه $ω_n$ بزرگ‌تر باشد، پاسخ ضربه فشرده‌تر شده، فرکانس نوسان‌های $h(t)$ و $s(t)$ بزرگ‌تر می‌شود. در واقع از معادله (۶-۴۱) چنین برمی‌آید که فرکانس نوسان $h(t)$ و $s(t)$ برابر $ω_n \sqrt{۱-ζ^۲}$ است، که با افزایش $ω_n$ زیاد می‌شود. البته این فرکانس صراحتاً به نسبت میرایی هم بستگی دارد و همیشه از $ω_n$ کوچک‌تر است، مگر در حالت نامیرا $ζ=۰$ (به همین خاطر است که $ω_n$ را فرکانس طبیعی نامیرا می‌نامند.)
در مثال فنر–ضربه‌گیر نتیجه می‌گیریم که در صورت نبودن کمک‌فنر، فرکانس نوسان‌های جرم $ω_n$ است و بودن کمک‌فنر این فرکانس را کمتر می‌کند.
</div>
<div dir="rtl" style="text-align: right;">
در شکل ۶-۲۳ نمودار بودی پاسخ فرکانسی معادله (۶-۳۳) به‌ازای چند مقدار $ζ$ رسم شده است. مثل حالت مرتبه اول، مقیاس لگاریتمی فرکانس باعث می‌شود که لگاریتم دامنه در فرکانس‌های پایین و بالا، مجانب‌های خط راست داشته باشد. از معادله (۶-۳۳) به‌دست می‌آوریم:
</div>

$$
۲۰ \log_{۱۰} |H(jω)| = -۱۰ \log_{۱۰} \left\{ [۱ - (ω / ω_n)^۲]^۲ + ۴ζ^۲ (ω / ω_n)^۲ \right\}
\tag{۶-۴۳}
$$
<div dir="rtl" style="text-align: right;">
که به‌دست می‌دهد:
</div>

$$
۲۰ \log_{۱۰} |H(jω)| ≈
\begin{cases}
۰ & \text{ω ≪ ωₙ} \\
-۴۰ \log_{۱۰} ω + ۴۰ \log_{۱۰} ω_n & \text{ω ≫ ωₙ}
\end{cases}
\tag{۶-۴۴}
$$

<div style="text-align: center;">
    <img src="https://raw.githubusercontent.com/tayebehemati/DSP/439a15d2d250455385e2603470782465424873d7/images/%D8%B4%DA%A9%D9%846-23.jpg"
         alt="شکل ۶-۲۲"
         style="display: block; margin: auto; width: 50%;">  
<p style="font-size: 10pt; text-align: center;">
شکل ۶-۲۳ نمودار بودی سیستم مرتبه دوم به ازای چند نسبت میرایی $ζ$ مختلف
</p>

```python
# تعریف پاسخ فرکانسی H(jω) برای سیستم مرتبه دوم
def magnitude_response(omega, zeta, omega_n):
    return 1 / np.sqrt((1 - (omega / omega_n)**2)**2 + (2 * zeta * omega / omega_n)**2)

# بردار فرکانس‌ها در مقیاس لگاریتمی
omega = np.logspace(-1, 2, 1000) * omega_n

# رسم نمودار بهره فرکانسی در مقیاس dB
plt.figure(figsize=(10, 6))
for zeta in zeta_values:
    H = magnitude_response(omega, zeta, omega_n)
    plt.plot(omega / omega_n, 20 * np.log10(H), label=f'ζ = {zeta}')

# تنظیمات ظاهری نمودار
plt.xscale('log')
plt.ylim([-80, 20])
plt.xlabel(r'$\omega$')
plt.ylabel(r'$20 \log_{10}|H(j\omega)|$')
plt.title('')
plt.xticks([0.1, 1, 10, 100], [r'$0.1\omega_n$', r'$\omega_n$', r'$10\omega_n$', r'$100\omega_n$'])
plt.axhline(0, color='gray', linewidth=0.5)
plt.grid(True, which='both', ls='--', lw=0.5)
plt.legend()
plt.tight_layout()
plt.show()

import numpy as np
import matplotlib.pyplot as plt

# پارامترهای اصلی
omega_n = 1  # فرکانس طبیعی
omega = np.logspace(-1, 2, 500)  # محور فرکانسی لگاریتمی از 0.1 تا 100
zeta_values = [0.1, 0.2, 0.4, 0.7, 1, 1.5]  # مقادیر مختلف میرایی

# تعریف تابع فاز پاسخ فرکانسی
def phase_response(omega, zeta, omega_n):
    numerator = 2 * zeta * omega / omega_n
    denominator = 1 - (omega / omega_n)**2
    return -np.arctan2(numerator, denominator)

# رسم نمودار
plt.figure(figsize=(10, 6))
for zeta in zeta_values:
    phase = phase_response(omega, zeta, omega_n)
    plt.plot(omega / omega_n, phase, label=f'ζ = {zeta}')

# تنظیمات نمایشی
plt.xscale('log')
plt.xticks([0.1, 1, 10, 100], [r'$0.1\omega_n$', r'$\omega_n$', r'$10\omega_n$', r'$100\omega_n$'])
plt.yticks(
    [-np.pi, -3*np.pi/4, -np.pi/2, -np.pi/4, 0],
    [r'$-\pi$', r'$-\frac{3\pi}{4}$', r'$-\frac{\pi}{2}$', r'$-\frac{\pi}{4}$', r'$0$']
)
plt.xlabel(r'$\omega$')
plt.ylabel(r'$\angle H(j\omega)$')
plt.grid(True, which='both', ls='--', lw=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

<div dir="rtl" style="text-align: right;">
مجانب فرکانس پایین لگاریتم دامنه خط dB و مجانب فرکانس بالای آن، طبق معادله (۶-۴۴)، خطی با شیب ۴۰dB-  بر دهه است. یعنی به ازای هر ده برابر شدن فرکانس، $\left|H(j\omega)\right|$ به اندازه ۴۰dB  افت می‌کند. به علاوه  دو مجانب خطی هم را در $\omega = \omega_n$ قطع میکنند. پس با استفاده از تقریب معادله (۶-۴۴) می توانیم برای لگاریتم  دامنه یک تقریب تکه تکه خطی به دست آوریم. $\omega_n$ محل تغییر شیب این منحنی تقریبی است و به همین خاطر  آن را فرکانس شکست سیستم مرتبه دوم مینامند. 
در شکل ۶-۳۳ منحنی تقریبی به صورت خط چین رسم شده است.
</div>
<div dir="rtl" style="text-align: right;">
برای $\angle H(j\omega)$ نیز میتوان تقریب خط راست به دست آورد مقدار دقیق $\angle H(j\omega)$ را میتوان از معادله  (۳۳-۶) یافت:  
</div>

$$
\angle H(j\omega) = -\tan^{-۱} \left( \frac{۲\zeta(\omega∕\omega_n )}{۱-(\omega⁄\omega_n )^۲ } \right)
\tag{۶-۴۵}
$$
<div dir="rtl" style="text-align: right;">
تقریب عبارت است از 
</div>

$$
\angle H(j\omega) \approx 
\begin{cases}
۰ & \omega \leq {۰/۱} \omega_n \\
-\frac{\pi}{۲} \left[ \log_{۱۰}(\omega/\omega_n) + ۱ \right] & {۰/۱} \omega_n \leq \omega \leq ۱۰\omega_n \\
-\pi & \omega \geq ۱۰\omega_n
\end{cases}
\tag{۶-۴۶}
$$
<div dir="rtl" style="text-align: right;">
که در شکل ۶-۲۳ رسم شده است. باز هم مقدار واقعی و مقدار تقریبی در $\omega = \omega_n$ برابرند، در این فرکانس 
</div>

$$
\angle H(j\omega_n ) = -\frac{\pi}{۲}
$$
<div dir="rtl" style="text-align: right;">
توجه به این نکته مهم است که تقریبهای مجانبی به دست آمده برای سیستم مرتبه دوم به $\zeta$ بستگی ندارند، حال  آنکه نمودارهای واقعی $\left|H(j\omega)\right|$ و $\angle H(j\omega)$ مسلماً به $\zeta$ وابسته اند و برای رسم دقیقتر این نمودارها، مخصوصاً  نزدیک فرکانس قطع $\omega = \omega_n$، باید این مطلب را در نظر گرفت و تقریب را اصلاح کرد تا به نمودار واقعی  شبیه تر شود. تفاوت منحنی های تقریبی و واقعی به ازای $\zeta$ کوچک آشکار تر است. مخصوصاً دقت کنید که در  حوالی $\omega = \omega_n$ منحنی واقعی لگاریتم دامنه قله دارد. محاسبه ای سر راست، به کمک معادله (۶-۴۳) نشان  می دهد که به ازای $\zeta < \sqrt{۲/۲} \approx ۰.۷۰۷$، $\left|H(j\omega)\right|$ در فرکانس زیر ماکزیمم می شود.
</div>

$$
\omega_\text{max} = \omega_n \sqrt{۱ - ۲\zeta^۲}
\tag{۶-۴۷}
$$
<div dir="rtl" style="text-align: right;">
و مقدار این ماکزیمم عبارت است از  
</div>

$$
\left|H(j\omega_\text{max})\right| = \frac{۱}{۲\zeta \sqrt{۱ - \zeta^{۲}}}
\tag{۶-۴۸}
$$
<div dir="rtl" style="text-align: right;">
ولی به ازای $\zeta > ۰.۷۰۷$، $H(j\omega)$ با افزایش $\omega$ به طور یکنوا کاهش می یابد. اینکه $H(j\omega)$ می تواند قله داشته باشد  در طراحی فیلترها و تقویت کننده های فرکانس گزین بسیار اهمیت دارد. در بعضی کاربردها طراحی می خواهد  مداری طرح کند که اندازه پاسخ فرکانسی آن در فرکانس خاصی قله تیزی داشته باشد، تا تقویت به صورت  فرکانس گزین انجام شود و تنها فرکانسهای واقع در یک باند باریک تقویت شوند. ضریب کیفیت $Q$ معیاری از  تیزی این قله است. برای یک مدار مرتبه دوم توصیف شده با معادله ای به شکل معادله (۶-۳۱)، کیفیت را  معمولاً چنین تعریف می کنند.  
</div>

$$
Q = \frac{۱}{۲\zeta}
$$
<div dir="rtl" style="text-align: right;">
شکل ۶-۲۳ و معادله (۶-۴۸) نشان میدهند که این تعریف درست است هر چه میرایی کمتر باشد قله $\left|H(j\omega)\right|$  تیز تر است.

</div>
