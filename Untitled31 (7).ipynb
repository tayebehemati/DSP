<div dir="rtl" style="text-align: right;">
و پاسخ ضربه آن عبارت است از  
</div>
          
$$
h(t) = \frac{۱}{\tau} e^{\left(\frac{-t}{\tau}\right)} u(t) \tag {۶-۲۳}
$$ 
<div dir="rtl" style="text-align: right;">
که در شکل ۶-۱۹ (الف) رسم شده است. پاسخ پله سیستم عبارت است از  
</div>

$$
s(t) = h(t) * u(t) = \left[۱ - e^{\left(\frac{-t}{\tau}\right)}\right] u(t) \tag{۶-۲۴}
$$  
<div dir="rtl" style="text-align: right;">
که در شکل ۶-۱۹ (ب) رسم شده است. پارامتر $\tau$ ثابت زمانی سیستم است و آهنگ پاسخ سیستم مرتبه اول را کنترل می‌کند.  
برای مثال، چنانچه شکل ۶-۱۹ نشان می‌دهد، در $t = \tau$ پاسخ ضربه به $۱/e$ اندازه پاسخ در $t = ۰$، و پاسخ پله به $۱/e$ مقدار نهایی‌اش می‌رسد. بنابراین هر چه $\tau$ کوچکتر باشد، پاسخ ضربه تندتر نزول می‌کند و زمان صعود پاسخ پله کمتر می‌شود؛ یعنی پاسخ پله تندتر به مقدار نهایی خود می‌رسد. همچنین توجه کنید که پاسخ پله سیستم مرتبه اول نوسان میرا ندارد.
</div>

<img src="https://raw.githubusercontent.com/sepidehrezvani/page-404-408-/ba2ea6afbf42b0ead046dae574f91325f9b77d33/%D8%B4%DA%A9%D9%84%206-19%20%D8%B5%D9%81%D8%AD%D9%87%20406.png" style="width:50%; display:block; margin:auto;">
<p style="text-align:center; font-size:12px;">شکل ۶-۱۹ - پاسخ پله سیستم مرتبه اول</p>
    


```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیمات
tau = 1
t = np.linspace(0, 5*tau, 500)

# (الف) پاسخ ضربه h(t) = (1/τ) * exp(-t/τ)
h = (1/tau) * np.exp(-t/tau)

# (ب) پاسخ پله s(t) = 1 - exp(-t/τ)
s = 1 - np.exp(-t/tau)

# رسم نمودار
fig, axs = plt.subplots(2, 1, figsize=(5, 6))

# نمودار (الف)
axs[0].plot(t, h, 'k', linewidth=2)
axs[0].hlines(1/tau, 0, 0.1, colors='k')
axs[0].hlines(1/(tau*np.e), 0, tau, colors='k', linestyles='dashed')
axs[0].axvline(tau, color='k', linestyle='dashed')
axs[0].text(0.02, 1/tau + 0.05, r'$1/\tau$', fontsize=12)
axs[0].text(tau + 0.1, 1/(tau*np.e), r'$1/(\tau e)$', fontsize=12)
axs[0].text(tau - 0.05, -0.02, r'$\tau$', fontsize=12)
axs[0].set_ylabel('h(t)', fontsize=14)
axs[0].set_xticks([])
axs[0].set_yticks([])
axs[0].set_xlim(0, 5*tau)
axs[0].set_ylim(0, 1.2/tau)
axs[0].set_title('(الف)', loc='right', fontsize=14)

# نمودار (ب)
axs[1].plot(t, s, 'k', linewidth=2)
axs[1].hlines(1, 0, 0.1, colors='k', linestyles='dashed')
axs[1].hlines(1 - 1/np.e, 0, tau, colors='k', linestyles='dashed')
axs[1].axvline(tau, color='k', linestyle='dashed')
axs[1].text(0.05, 1.02, '1', fontsize=12)
axs[1].text(tau + 0.1, 1 - 1/np.e, r'$1 - \frac{1}{e}$', fontsize=12)
axs[1].text(tau - 0.05, -0.05, r'$\tau$', fontsize=12)
axs[1].set_ylabel('s(t)', fontsize=14)
axs[1].set_xlabel('t', fontsize=14)
axs[1].set_xticks([])
axs[1].set_yticks([])
axs[1].set_xlim(0, 5*tau)
axs[1].set_ylim(0, 1.1)
axs[1].set_title('(ب)', loc='right', fontsize=14)

plt.tight_layout()
plt.show()
```

<div dir="rtl" style="text-align: right;">
شکل ۶-۲۰ نمودار بودی پاسخ فرکانسی معادله (۶-۲۲) را نشان می‌دهد. در این شکل یک مزیت به کار بردن مقیاس لگاریتمی برای فرکانس نشان داده شده است. می‌توانیم بدون زحمت زیاد تقریب مفیدی از نمودار سیستم مرتبه اول به دست آوریم.  
برای پی بردن به این مطلب ابتدا نمودار لگاریتم دامنه پاسخ فرکانسی را بررسی می‌کنیم. از معادله (۶-۲۲) به دست می‌آوریم  
</div>
 
$$
۲۰ \log_{۱۰} |H(j\omega)| = -۱۰ \log_{۱۰} \left[(\omega \tau)^۲ + ۱\right] \tag {۶-۲۵}
$$  
<div dir="rtl" style="text-align: right;">
می‌بینیم که به ازای $\omega\tau \ll ۱$، لگاریتم دامنه تقریباً صفر است، در حالی که به ازای $\omega\tau \gg ۱$ لگاریتم دامنه تقریباً تابعی خطی از $\log_{۱۰} (\omega)$ است. یعنی  
</div>

<div dir="rtl" style="text-align: right;">
به ازای  
</div>

$$
۲۰ \log_{۱۰} |H(j\omega)| \approx ۰ \qquad \omega \ll \frac{۱}{\tau} \tag {۶-۲۶}
$$  

<div dir="rtl" style="text-align: right;">
و به ازای   
</div>
    
$$
۲۰ \log_{۱۰} |H(j\omega)| \approx -۲۰ \log_{۱۰} (\omega \tau) = -۲۰ \log_{۱۰} (\omega) - ۲۰ \log_{۱۰} (\tau) \qquad \omega \gg \frac{۱}{\tau} \tag{۶-۲۷}
$$

<img src="https://raw.githubusercontent.com/sepidehrezvani/page-404-408-/ba2ea6afbf42b0ead046dae574f91325f9b77d33/%D8%B4%DA%A9%D9%84%206-20%20%D8%B5%D9%81%D8%AD%D9%87%20407.png" style="width:50%; display:block; margin:auto;">
<p style="text-align:center; font-size:12px;">.شکل ۶-۲۰ نمودار بودی سیستم مرتبه اول پیوسته در زمان</p>



```python
import numpy as np
import matplotlib.pyplot as plt

# محدوده فرکانس
w = np.logspace(-1, 2, 500)  # از 0.1 تا 100 (برحسب 1/τ)

# تابع انتقال فیلتر پایین‌گذر درجه اول
H = 1 / np.sqrt(1 + w**2)

# پاسخ دسی‌بل
H_db = 20 * np.log10(H)

# تقریب مایانی: برای w > 1/τ => H ~ 1/w
w_approx = w[w >= 1]
H_approx_db = -20 * np.log10(w_approx)

# رسم
plt.figure(figsize=(6, 3))
plt.semilogx(w, H_db, 'k', linewidth=2, label=r'$|H(j\omega)|$')
plt.semilogx(w_approx, H_approx_db, 'k--', linewidth=1.5, label='تقریب میانی')

# خطوط راهنما برای 3dB
plt.plot([1, 1], [-3, 0], 'k:', linewidth=1)
plt.plot([0.9, 1.1], [-3, -3], 'k:', linewidth=1)
plt.text(1.05, -2, '3 dB', fontsize=12)

# تنظیمات محور و برچسب‌ها
plt.ylim(-60, 10)
plt.xlim(0.1, 100)
plt.xticks([0.1, 1, 10, 100], [r'$0.1/\tau$', r'$1/\tau$', r'$10/\tau$', r'$100/\tau$'])
plt.yticks(np.arange(-60, 21, 20))
plt.xlabel(r'$\omega$', fontsize=14)
plt.ylabel(r'$20 \log_{10} |H(j\omega)|$', fontsize=14)
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.legend(loc='upper right', fontsize=11)
plt.tight_layout()
plt.show()

import numpy as np
import matplotlib.pyplot as plt

# پارامتر زمانی
tau = 1
omega = np.logspace(-1, 2, 500)  # از 0.1 تا 100 (در مقیاس لگاریتمی)

# تابع تبدیل: H(jω) = 1 / (1 + jωτ)
H = 1 / (1 + 1j * omega * tau)
phase = np.angle(H)  # فاز واقعی

# تقریب مداری (فاز تقریبی)
phase_approx = np.piecewise(
    omega,
    [omega < 1/tau, (omega >= 1/tau) & (omega <= 10/tau), omega > 10/tau],
    [0, lambda w: -np.pi/4 * np.log10(w * tau * 10), -np.pi/2]
)

# رسم نمودار
plt.figure(figsize=(6, 3))
plt.semilogx(omega/tau, phase, 'k', label='فاز دقیق', linewidth=2)  # نمودار اصلی
plt.semilogx(omega/tau, phase_approx, 'k--', label='تقریب مداری', linewidth=1.5)  # نمودار خط‌چین

# برچسب محور و مقیاس
plt.yticks(
    [-3*np.pi/4, -np.pi/2, -np.pi/4, 0, np.pi/4],
    [r'$-\frac{3\pi}{4}$', r'$-\frac{\pi}{2}$', r'$-\frac{\pi}{4}$', '0', r'$\frac{\pi}{4}$']
)
plt.xticks([0.1, 1, 10, 100], [r'$0.1/\tau$', r'$1/\tau$', r'$10/\tau$', r'$100/\tau$'])
plt.xlabel(r'$\omega$', fontsize=14)
plt.ylabel(r'$\angle H(j\omega)$', fontsize=14)

# متن فارسی
plt.text(0.15, -0.15, 'تقریب مجانی', fontsize=12)

plt.grid(True, which='both', linestyle='--', alpha=0.3)
plt.legend(loc='lower left')
plt.ylim(-3*np.pi/4, np.pi/4)
plt.xlim(0.08, 120)
plt.tight_layout()
plt.show()
```

<div dir="rtl" style="text-align: right;">
به بیان دیگر، برای سیستم درجه اول، مجانی‌های لگاریتم دامنه در فرکانس‌های پایین و بالا، خط راست‌اند. مجانب فرکانس پایین [معادله (۶-۲۶)] خط $۰dB$ و مجانب فرکانس بالا [معادله (۶-۲۷)] خطی است که به ازای هر دهه افزایش (یعنی ۱۰ برابر شدن آن)، $|H(j\omega)|$ به اندازه $۲۰dB$ افت می‌کند. این مجانب را گاهی مجانب «$۲۰dB$ بر دهه» می‌نامند.
</div>
<div dir="rtl" style="text-align: right;">
توجه کنید که دو تقریب مجانبی بیان شده با معادلات (۶-۲۶) و (۶-۲۷) در نقطه  
$\log_{۱۰}(j\omega) = -\log_{۱۰}(\tau)$  
با $\omega = \dfrac{۱}{\tau}$ مساوی‌اند. از لحاظ ترسیمی، دو مجانب همدیگر را در نقطه $\omega = \dfrac{۱}{\tau}$ قطع می‌کنند.  
تقریب مجانبی لگاریتم دامنه با دو خط راست با توجه به این مطلب به دست می‌آید.  
تقریب $۲۰\log_{۱۰}|H(j\omega)|$ با صفر در $\omega \ll \dfrac{۱}{\tau}$ و با معادله (۶-۲۷) در $\omega \gg \dfrac{۱}{\tau}$ داده می‌شود.  
این تقریب نیز در شکل ۶-۲۰ (با خط چین) نشان داده شده است.
نقطه‌ای که در آن شیب تقریب تغییر می‌کند، دقیقاً $\omega = \dfrac{۱}{\tau}$ است که به همین دلیل آن را <b>فرکانس شکست</b> می‌نامند.  
همچنین توجه کنید که در $\omega = \dfrac{۱}{\tau}$، دو جملهٔ آرگومان لگاریتم معادله (۶-۲۵) برابرند.  
پس در این نقطه مقدار واقعی اندازه عبارت است از
</div>

$$
۲۰\log_{۱۰}|H(j\dfrac{۱}{\tau})| = -۱۰\log_{۱۰}(۲) \approx -۳\,dB
$$

<div dir="rtl" style="text-align: right;">
به همین خاطر گاهی نقطه $\omega = \dfrac{۱}{\tau}$ را نقطه dB۳ می‌نامند.  
شکل نشان می‌دهد که نمودار بودی تنها در حوالی فرکانس شکست با مجانب خط راست تفاوت زیادی دارد.  
بنابراین برای ترسیم دقیق‌تر نمودار بودی، تنها باید مجانب‌ها را نزدیک فرکانس شکست اصلاح کنیم.
برای $\angle H(j\omega)$ نیز می‌توان تقریب خط راست مفیدی به دست آورد
</div>

$$
\angle H(j\omega) = -\tan^{-۱}(\omega\tau) \approx 
\begin{cases}
۰ & \omega \leq \dfrac{۰٫۱}{\tau} \\
-\dfrac{\pi}{۴}[\log_{۱۰}(\omega\tau)+۱] & \dfrac{۰٫۱}{\tau} \leq \omega \leq \dfrac{۱۰}{\tau} \\
-\dfrac{\pi}{۲} & \omega \geq \dfrac{۱۰}{\tau}
\end{cases}
$$
<div dir="rtl" style="text-align: right;">
توجه کنید که در بازه $\dfrac{۰٫۱}{\tau} \leq \omega \leq \dfrac{۱۰}{\tau}$،  
مقدار این تقریب (از صفر تا $-\dfrac{\pi}{۲}$) به‌صورت خطی کاهش می‌یابد.  
مجانب واقعی $\angle H(j\omega)$ به ازای $\omega \ll \dfrac{۱}{\tau}$ برابر $۰$  
و به ازای $\omega \gg \dfrac{۱}{\tau}$ برابر $-\dfrac{\pi}{۲}$ است.  
همچنین تقریب در فرکانس شکست $\omega = \dfrac{۱}{\tau}$ با مقدار واقعی برابر است. در این نقطه داریم:
</div>

$$
 \angle H(j\omega)(j\dfrac{۱}{\tau}) = -\dfrac{\pi}{۴} \tag{۴-۱۶۸}
$$

<div dir="rtl" style="text-align: right;">
تقریب مجانبی فاز نیز در شکل ۶-۲۰ رسم شده است و این شکل، چگونگی اصلاح مجانب‌های راست برای دقیق‌تر شدن منحنی $\angle H(j\omega)$ را نیز نشان می‌دهد.
با توجه به این سیستم مرتبه اول، بار دیگر می‌توان رابطه معکوس بین حوزه‌های زمان و فرکانس را مشاهده کرد.  
هرچه $\tau$ را کوچکتر کنیم، پاسخ زمانی سیستم سریع‌تر می‌شود (یعنی $h(t)$ به طرف مبدأ فشرده‌تر، و زمان صعود پاسخ پله کوچکتر می‌شود)  
و به‌طور هم‌زمان فرکانس شکست بزرگ‌تر می‌شود (یعنی $H(j\omega)$ بهتر می‌شود زیرا روی گستره فرکانس بزرگ‌تری $|H(j\omega)| \approx ۱$ است).  
با ضرب پاسخ ضربه در $\tau$ و مقایسه $\tau h(t)$ و $H(j\omega)$ هم می‌توان این نکته را دریافت:
</div>

$$
\tau h(t) = e^{-\dfrac{t}{\tau}} u(t)
$$

$$
H(j\omega) = \dfrac{۱}{j\omega\tau + ۱}
$$

<div dir="rtl" style="text-align: right;">
به این ترتیب $\tau h(t)$ تابعی از $\dfrac{t}{\tau}$ و $H(j\omega)$ تابعی از $\omega\tau$ می‌شود.  
بنابراین تغییر با تغییر مقیاس زمان و فرکانس متناظر است.
</div>


<div dir="rtl" style="text-align: right;">
    
#### مثال ۶-۳

<div dir="rtl" style="text-align: right;">
دو فیلتر پایین‌گذر طراحی‌شده برای داشتن فرکانس قطع ۵۰۰Hz را در نظر بگیرید. هر دو فیلتر دارای پاسخ فرکانسی مرتبه پنجم گویا و پاسخ ضربه‌ای حقیقی هستند. این دو فیلتر با نام‌های فیلتر باترورث و فیلتر بیضوی شناخته می‌شوند که در عمل بسیار مورد استفاده قرار می‌گیرند.
</div>

<div dir="rtl" style="text-align: right;">
شکل ۶-۱۸ (الف)، اندازه پاسخ فرکانسی این دو فیلتر را نشان می‌دهد (محور فرکانس بر حسب هرتز است). باند گذار، ناحیه‌ای پیرامون فرکانس قطع (۵۰۰Hz) است که در آن، اندازه پاسخ فرکانسی از مقدار یک، بیش از $۰/۰۵$ (تموج باند عبور) و از مقدار صفر نیز بیش از $۰/۰۵$ (تموج باند حذف) فاصله دارد.  
از شکل ۶-۱۸ (الف) می‌توان نتیجه گرفت که باند گذار فیلتر باترورث عریض‌تر از فیلتر بیضوی است.
بهایی که در ازای باریکی باند گذار در فیلتر بیضوی پرداخت شده، در شکل ۶-۱۸ (ب) مشخص است؛ این شکل پاسخ پله دو فیلتر را نشان می‌دهد. دیده می‌شود که نوسان‌های میرای پاسخ پله فیلتر بیضوی شدیدتر از فیلتر باترورث است. به‌ویژه، زمان نشست پاسخ پله در فیلتر بیضوی طولانی‌تر است.
</div>

# ۵-۶ سیستم‌های پیوسته در زمان مرتبه اول و دوم

<div dir="rtl" style="text-align: right;">
سیستم‌های LTI که با معادلات دیفرانسیل خطی با ضرایب ثابت توصیف می‌شوند، کاربرد فراوانی دارند؛ زیرا بسیاری از سیستم‌های فیزیکی را می‌توان با این معادلات مدل‌سازی کرد و پیاده‌سازی آن‌ها نیز ساده است.
به دلایل عملی، سیستم‌های مرتبه بالاتر معمولاً به‌صورت ترکیب‌های سری یا موازی از سیستم‌های مرتبه اول و دوم ساخته می‌شوند. بنابراین، شناخت دقیق سیستم‌های مرتبه اول و دوم برای تحلیل و طراحی سیستم‌های پیچیده‌تر ضروری است.
در این بخش، به بررسی سیستم‌های مرتبه پایین می‌پردازیم. همچنین در بخش ۶-۶، معادل‌های گسسته‌زمان این سیستم‌ها معرفی خواهند شد.
</div>

### ۶-۵-۶ سیستم‌های درجه اول

<div dir="rtl" style="text-align: right;">
معادله دیفرانسیل سیستم درجه اول معمولاً به صورت زیر نوشته می‌شود:
</div>

$$
r \frac{\mathrm{d}y(t)}{\mathrm{d}t} + y(t) = x(t)
$$
<div dir="rtl" style="text-align: right;">
در این رابطه، $r$ ضریبی است که اهمیت آن بعداً مشخص خواهد شد.
پاسخ فرکانسی سیستم درجه اول به صورت زیر است:
</div>

$$ H(jω)=\frac{۱}{jωτ+۱}$$

         alt="شکل ۶-۱۸"
         style="display: block; margin: auto; width: 50%;"> 
<p style="font-size:10px; text-align:center;">
شکل ۶-۱۸ نمونه هایی از فیلتر با ترورث و بیضوی مرتبه پنجم که تموج باند عبور ، تموج باند حذف ، و فرکانس قطع  یکسانی دارند: (الف) اندازه های پاسخ فرکانسی بر حسب فرکانس فرکانس بر حسب هرتز ؛ (ب) پاسخ پله.  
</p>
