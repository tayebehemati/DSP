```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(10, 4))
ax.set_xlim(-2, 10)
ax.set_ylim(-2, 3)
ax.axis('off')

# دیوار
wall = patches.Rectangle((-2, -1), 0.3, 4, hatch='///', color='lightgray')
ax.add_patch(wall)

# فنر (به صورت زیگ‌زاگ دستی)
x0, y0 = [-1.7, -1.5, -1.3, -1.1, -0.9, -0.7, -0.5, -0.3, -0.1, 0.1, 0.3], [1, 1.4, 0.6, 1.4, 0.6, 1.4, 0.6, 1.4, 0.6, 1.4, 1]
ax.plot(x0, y0, color='black', lw=2)
ax.text(-1.2, 1.6, 'k', fontsize=14)

# دمپر
ax.plot([-1.7, -1.3], [0, 0], color='black', lw=2)
ax.add_patch(patches.Rectangle((-1.3, -0.3), 0.4, 0.6, fill=False))
ax.plot([-0.9, -0.5], [0, 0], color='black', lw=2)
ax.text(-1.1, -0.5, 'b', fontsize=14)
ax.text(-1.15, -0.9, 'ضریب گیر', fontsize=12)

# جرم (مربع متحرک)
ax.add_patch(patches.Rectangle((0, -0.5), 2, 2, fill=False, lw=2))
ax.text(1, 0.5, 'm', fontsize=16, ha='center')

# چرخ‌ها
circle1 = patches.Circle((0.5, -0.5), 0.3, fill=False, lw=2)
circle2 = patches.Circle((1.5, -0.5), 0.3, fill=False, lw=2)
ax.add_patch(circle1)
ax.add_patch(circle2)

# نیروی اعمالی
ax.annotate('', xy=(3.5, 0.5), xytext=(2, 0.5), arrowprops=dict(arrowstyle='->', lw=2))
ax.text(3.6, 0.5, 'x(t)', fontsize=14)
ax.text(3.6, 0.2, '(نیروی اعمالی)', fontsize=12)

# پاسخ
ax.annotate('', xy=(3, 1.5), xytext=(2, 1.5), arrowprops=dict(arrowstyle='->', lw=2))
ax.text(3.2, 1.5, 'y(t)', fontsize=14)
ax.text(3.2, 1.8, '(پاسخ)', fontsize=12)

# جهت بیرونی
ax.annotate('', xy=(2, -1.5), xytext=(0, -1.5), arrowprops=dict(arrowstyle='->', lw=2))
ax.text(1, -1.8, '(بیرونی اعمالی)', fontsize=12, ha='center')

plt.show()
```
