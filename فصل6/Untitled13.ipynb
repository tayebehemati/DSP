## ۶-۲-۱ فاز خطی و غیر خطی

اگر تغییر فاز تابعی خطی از $\omega$ باشد، اثر آن تعبیر سرراستی در حوزه زمان دارد. سیستم LTI پیوسته در زمانی با پاسخ فرکانسی زیر را در نظر بگیرید:  

$$
H(j\omega) = e^{-j\omega t_0}
$$  

پس بهره سیستم یک و فاز آن خطی است، یعنی:  

$$
\angle H(j\omega) = -\omega t_0, \quad |H(j\omega)| = 1
$$  

در مثال ۴-۱۵ نشان دادیم که در سیستمی با این پاسخ فرکانسی، خروجی جابجا شده زمانی ورودی است، یعنی:  

$$
y(t) = x(t - t_0)
$$  

در حالت گسسته در زمان، فاز خطی اثری مشابه حالت پیوسته در زمان دارد، به شرطی که شیب فاز خطی عدد صحیح باشد. از مثال ۵-۱۱ به یاد داریم که سیستم LTI دارای پاسخ فرکانسی $e^{-j\omega n_0}$ تابع فاز خطی دارد و خروجی آن جابجا شده ورودی آن است، یعنی:  

$$
y[n] = x[n - n_0]
$$  

پس فاز خطی دارای شیب صحیح با جابجایی $x[n]$ به اندازه آن عدد صحیح متناظر است. اگر شیب فاز عدد صحیح نباشد، اثری که در حوزه زمان ایجاد می‌شود پیچیده‌تر است و در فصل ۷ بخش ۷-۵ مورد بحث قرار خواهد گرفت.  

به طور حسی می‌توان گفت که پوش مقادیر سیگنال جابجا می‌شود ولی خود مقادیر ممکن است تغییر کنند.  

تغییر فاز خطی به تغییر ساده و ملموسی در حوزه زمان منجر می‌شود، ولی اگر ورودی در معرض تغییر فاز غیر خطی قرار گیرد، مولفه‌های مختلف ورودی به نحوی تغییر می‌کنند که فازهای نسبی‌شان عوض می‌شود. با بر نهادن این نمایی‌ها سیگنالی به دست می‌آید که منظری کاملاً متفاوت نسبت به سیگنال ورودی دارد.

**شکل ۶-۳**  
(الف) سیگنال پیوسته در زمانی که به عنوان ورودی چند سیستم با پاسخ فرکانسی دارای بهره واحد به کار رفته است؛  
(ب) خروجی سیستم دارای فاز خطی؛  
(ج) خروجی سیستم دارای فاز غیر خطی؛  
(د) پاسخ سیستمی با همان تابع فاز سیستم (ج) به اضافه یک فاز خطی.  

سیستم دیگری با فاز غیر خطی که با افزودن یک فاز خطی به سیستم قبلی به دست آمده است، یعنی:  

$$
H_3(j\omega) = H_2(j\omega) \cdot e^{-j\omega n_0}
$$  

که در آن:  

$$
H_2(j\omega) = e^{j\angle H_2(j\omega)}
$$  

تابعی غیر خطی از $\omega$ است.  

پس خروجی شکل (د) را می‌توان پاسخ اتصال متوالی سیستم $H_2(j\omega)$ و سیستمی با فاز خطی دانست، به نحوی که سیگنال‌های شکل‌های (ج) و (د) تنها جابجا شده یکدیگر باشند.

**شکل ۶-۴**  
(الف) سیگنال گسسته در زمانی که به چند سیستم با اندازه پاسخ فرکانسی یک اعمال شده است؛  
(ب) پاسخ سیستم دارای فاز خطی با شیب ۵-؛  
(ج) پاسخ سیستم با فاز غیر خطی؛  
(د) پاسخ سیستمی که همان فاز غیر خطی شکل (ج) را دارد ولی یک فاز خطی نیز با آن جمع شده است.  

توجه کنید که تمام سیستم‌های در نظر گرفته شده در شکل‌های ۶-۳ و ۶-۴ بهره‌ای واحد دارند، یعنی اندازه تبدیل فوریه ورودی بدون تغییر در خروجی ظاهر می‌شود. به همین خاطر این سیستم‌ها را تمام‌گذر می‌نامند.  

مشخصات یک سیستم تمام‌گذر به طور کامل توسط جابجایی فاز آن تعیین می‌شود. البته در حالت کلی سیستم‌های LTI هم اندازه تبدیل فوریه را با بهره $|H(j\omega)|$ یا $|H(e^{j\omega})|$ تغییر می‌دهند و هم فاز آن را، و این تغییر فاز می‌تواند خطی یا غیر خطی باشد.
