# ۱-۱- مقدمه‌ای بر سیستم‌های کنترل

اتوماسيون واژه‌ای برای توصیف عملیات خودکار و یا کنترل فرآیند می‌باشد که در این تعریف **سیستم کنترل خودکار** ترکیبی از اجزاء مختلف می‌باشد که رفتار آنها به گونه‌ای است که کل سیستم را به‌صورت خودکار به هدف از پیش تعیین‌شده و مورد نظر هدایت می‌کند.

یک بررسی نزدیک از دستگاه‌ها و ماشین‌های تولیدشده امروزی پیرامون خودمان منجر به این نتیجه‌گیری می‌شود که بخشی یا کل این دستگاه‌ها و ماشین‌ها به‌صورت خودکار عمل می‌کنند.

## مثال‌هایی از سیستم‌های کنترل خودکار

برای مثال می‌توان به موارد زیر اشاره کرد:

- **لوازم خانگی:** یخچال و فریزر، آبگرمکن، ماشین لباسشویی، آسانسور، کنترل تلویزیون  
- **سیستم‌های ارتباطی و اینترنت**  
- **صنایع مختلف:** صنایع غذایی، کاغذسازی، سیمان و خودرو که تا حدی یا کاملاً به‌صورت خودکار اداره می‌شوند.  
- **نیروگاه‌ها:** نیروگاه‌های تولید برق، راکتورهای هسته‌ای و شیمیایی  
- **حمل و نقل:** سیستم‌های حمل و نقل خودرو، هواپیما، هلیکوپتر، زیردریایی  
- **ربات‌ها:** ربات‌های مونتاژ، جوش و غیره  
- **سیستم‌های نظامی:** کنترل آتش، پرتاب موشک و غیره  
- **کامپیوتر و تجهیزات جانبی:** چاپگرها، دیسک‌خوان‌ها، دیسک‌های مغناطیسی و غیره  
- **کشاورزی:** گلخانه، آبیاری و غیره  
- **سایر کاربردها:** کنترل موقعیت یا سرعت، دما، فشار، ولتاژ، سطح آب، ترافیک، مدیریت انرژی برای ساختمان  

## اهمیت سیستم‌های کنترل خودکار

تمامی مثال‌های فوق به این نتیجه مهم منجر می‌شوند که **کنترل اتوماتیک با همه زمینه‌های زندگی انسان عجین شده است**. مشخصه متمایز سیستم‌های کنترل اتوماتیک نسبت به سیستم‌های دیگر این است که با کاهش مشارکت انسانی در فعالیت‌های پرزحمت، خطرناک و مضر، از طرفی سبب افزایش تولید، بهبود کیفیت و ارزان‌تر شدن فرآورده‌ها شده‌اند.
