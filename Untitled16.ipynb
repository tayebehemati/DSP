# مشخصات زمانی و فرکانسی سیگنال‌ها و سیستم‌ها

## ۶-۰ مقدمه  

توصیف یک سیستم LTI بر حسب پاسخ فرکانسی جایگزینی برای توصیف حوزه زمان سیستم بر حسب کانولوشن است.  
در تحلیل سیستم‌های LTI غالباً استفاده از حوزه فرکانسی مناسب‌تر است زیرا معادلات دیفرانسیل و تفاضلی و عمل کانولوشن در حوزه فرکانس به عملیات جبری تبدیل می‌شوند.  
همچنین مفاهیمی چون فرکانس‌گزینی را در حوزه فرکانس می‌توان راحت تصور کرد.  

البته در طراحی سیستم معمولاً باید هم ملاحظات حوزه فرکانس منظور شود و هم ملاحظات حوزه زمان.  
برای نمونه همان‌طور که در مثال‌های ۴-۱۸ و ۵-۱۲ به اختصار گفتیم، و در این فصل به تفصیل خواهیم گفت، رفتار نوسانی شدید پاسخ ضربه فیلترهای فرکانس‌گزین ممکن است نامطلوب باشد و مجبور شویم فرکانس‌گزینی را تا حدی فدای دستیابی به مشخصات مطلوب حوزه زمان کنیم.  

در عمل وضعیت‌های این‌چنین قاعده هستند نه استثنا، زیرا در اکثر کاربردها قیدهایی برای سیستم هم در حوزه زمان و هم در حوزه فرکانس تعیین می‌کنیم که به خواسته‌های متضاد می‌انجامد.  

پس در طراحی و تحلیل سیستم دانستن روابط بین حوزه فرکانس و حوزه زمان و بده‌بستان‌های موجود مهم است.  
معرفی این مباحث و روابط مهم‌ترین هدف این فصل است.  

## ۶-۱ نمایش اندازه - فاز تبدیل فوریه  

تبدیل فوریه در حالت کلی مختلط است و همان‌طور که قبلاً گفته‌ایم می‌توان آن را با بخش‌های حقیقی و موهومی یا اندازه و فازش نشان داد.  
نمایش اندازه - فاز تبدیل فوریه پیوسته در زمان \$X(j\omega)\$ عبارت است از:  

$$
X(j\omega) = |X(j\omega)| e^{j \angle X(j\omega)}
$$

(معادله ۶-۱)

به‌نحوی مشابه، نمایش اندازه - فاز تبدیل فوریه گسسته در زمان \$X(j\omega)\$ عبارت است از:

$$
X(j\omega) = |X(j\omega)| e^{j \angle X(j\omega)}
$$

(معادله ۶-۲)

در بحث زیر راجع به نمایش‌های اندازه - فاز توجه عمده خود را روی حالت پیوسته در زمان متمرکز می‌کنیم.  
نکات اصلی را به حالت گسسته در زمان نیز می‌توان اعمال کرد.  

از معادله ترکیب تبدیل فوریه یعنی معادله (۴-۸) چنین بر می‌آید که می‌توان \$X(j\omega)\$ را تجزیه‌ی سیگنال \$x(t)\$ به مجموع نمایی‌های مختلط با فرکانس‌های مختلف دانست.  

در حقیقت چنان‌که در بخش ۴-۳-۷ گفتیم \$|X(j\omega)|^2\$ را می‌توان طیف چگالی انرژی \$x(t)\$ دانست.  
یعنی \$|X(j\omega)|^2 \, d\omega / 2\pi\$ را می‌توان انرژی واقع در باند فرکانسی بی‌نهایت کوچک بین \$\omega\$ و \$\omega + d\omega\$ فرض کرد.  

پس اندازهٔ \$|X(j\omega)|\$ محتوای فرکانسی سیگنال را نشان می‌دهد.  
یعنی \$|X(j\omega)|\$ اطلاعات مربوط به اندازه نسبی نمایی‌های مختلط تشکیل‌دهندهٔ \$x(t)\$ را در بردارد.  

مثلاً اگر در خارج یک باند فرکانسی کوچک حول صفر داشته باشیم \$|X(j\omega)| = 0\$، آنگاه \$x(t)\$ تنها نوسان‌های فرکانس پایین دارد.

ولی زاویه فاز \$\angle X(j\omega)\$ بر اندازهٔ مولفه‌های فرکانسی اثری ندارد، بلکه اطلاعات مربوط به فاز نسبی این نمایی‌ها را نشان می‌دهد.  
رابطهٔ فاز موجود در \$\angle X(j\omega)\$ اثر زیادی بر طبیعت سیگنال \$x(t)\$ دارد و نوعاً اطلاعات زیادی در مورد سیگنال دارد.  

در واقع بسته به تابع فاز می‌توانیم سیگنال‌هایی با ظاهر متفاوت به دست آوریم، حتی اگر تابع اندازه تغییر نکند.  

### به عنوان نمونه:

دوباره مثال شکل ۳-۳ را در نظر بگیرید.  
در اینجا کشتی با جمع آثار سه موج روبرو است، که هر کدام با یک سیگنال سینوسی مدل شده‌اند.  

اگر این سه سینوسی دامنهٔ ثابتی داشته باشند، بسته به فاز نسبی‌شان می‌توانند دامنه بسیار بزرگ یا کوچکی داشته باشند.  
پس فازها برای کشتی کاملاً مهم‌اند.  

### مثال دیگر:

سیگنال زیر را در نظر بگیرید:  

$$
x(t) = 1 + \frac{1}{2} \cos(2\pi t + \varphi_1) + \cos(4\pi t + \varphi_2) + \frac{2}{3} \cos(6\pi t + \varphi_3)
$$

(معادله ۶-۳)

شکل ۳-۴ این سیگنال را به‌ازای \$\varphi_1 = \varphi_2 = \varphi_3 = 0\$ نشان می‌داد.  
در شکل ۱-۶ این \$x(t)\$ را به‌ازای چند فاز مخالف برای سینوس‌ها نشان داده‌ایم.  

همان‌طور که از شکل برمی‌آید، می‌بینیم که سیگنال حاصل بر حسب مقادیر مختلف فاز نسبی چقدر می‌تواند متفاوت باشد.  

به‌طور کلی تغییر تابع فاز \$X(j\omega)\$ به تغییر مشخصات حوزه زمان سیگنال \$x(t)\$ می‌انجامد.  

در بعضی موارد اعوجاج فاز مهم است و در بعضی موارد نه.  

مثلاً یکی از خواص سیستم شنوایی انسان عدم حساسیت به فاز است.  
اگر تبدیل فوریهٔ سیگنال یک صدای تنها (مثلاً یک مصوت) اعوجاج فاز داشته باشد ولی اندازه‌اش تغییر نکند،  
قابلیت فهم آن تغییر زیادی نمی‌کند، گرچه شکل موج آن در حوزه زمان می‌تواند شکلی کاملاً دگرگون به خود بگیرد.  

هرچند اعوجاج ملایم فاز، نظیر اعوجاج تک‌صداها، قابلیت تشخیص را از بین نمی‌برد،  
ولی اعوجاج فاز شدید گفتار، مسلماً قابلیت تشخیص آن‌ها را از بین می‌برد.


## به عنوان مثالی بارز

فرض کنید $x(t)$ سیگنال ضبط‌شده‌ی بیان یک جمله است. اگر نوار وارونه پخش شود، سیگنال $x(-t)$ به دست می‌آید، و اثر متناظر در حوزه فرکانس، منفی شدن فاز است:  

$$
F\{x(-t)\} = X(-j\omega) = |X(j\omega)| e^{-j\angle X(j\omega)}
$$

![شکل ۶-۱](https://raw.githubusercontent.com/sepidehrezvani/s/189ee51ca7b09d098b0069fcd1ca443ce1332610/Capture_2025_02_28_15_10_52_397.png)

**شکل ۶-۱** سیگنال $x(t)$ معادله (۶-۳) به ازای چند مقدار مختلف $\varphi_1$، $\varphi_2$، $\varphi_3$:

- (الف) $\varphi_1 = \varphi_2 = \varphi_3 = 0$
- (ب) $\varphi_1 = 12 \, \text{rad}$، $\varphi_2 = 8 \, \text{rad}$، $\varphi_3 = 4 \, \text{rad}$
- (ج) $\varphi_1 = 6 \, \text{rad}$، $\varphi_2 = -2.7 \, \text{rad}$، $\varphi_3 = 0.93 \, \text{rad}$
- (د) $\varphi_1 = 1.2 \, \text{rad}$، $\varphi_2 = 4.1 \, \text{rad}$، $\varphi_3 = -7.02 \, \text{rad}$

یعنی اندازه طیف نوار وارونه با طیف سیگنال اصلی برابر است و تنها فاز آن تغییر کرده است. واضح است که این تغییر فاز اثر مهمی بر قابلیت فهم صدای ضبط‌شده دارد.  

بررسی تصاویر مثال دیگری است که اهمیت فاز را نشان می‌دهد. همانطور که در فصل ۳ به اختصار گفتیم، تصویر سیاه و سفید را می‌توان با سیگنال $x(t_1, t_2)$ نشان داد که $t_1$ مختصه‌ی افقی نقطه‌ای از تصویر، $t_2$ مختصه‌ی قائم آن نقطه، و $x(t_1, t_2)$ روشنی تصویر در $(t_1, t_2)$ است.  

تبدیل فوریه $X(j\omega_1, j\omega_2)$ تصویر، تجزیه‌ی $x(t_1, t_2)$ به‌صورت نمایی‌های $e^{j\omega_1 t_1} e^{j\omega_2 t_2}$ را نشان می‌دهد که تغییرات مکان $x(t_1, t_2)$ در فرکانس‌های مختلف را در بردارد.

## ۶-۲ نمایش اندازه - فاز پاسخ فرکانسی سیستم‌های LTI

طبق خاصیت کانولوشن، تبدیل فوریه پیوسته در زمان $Y(j\omega)$ خروجی یک سیستم LTI و تبدیل $X(j\omega)$ ورودی آن به‌صورت زیر به هم مربوط می‌شوند:  

$$
Y(j\omega) = H(j\omega) X(j\omega)
$$

که در آن $H(j\omega)$ پاسخ فرکانسی سیستم؛ یعنی تبدیل فوریه پاسخ ضربه سیستم است. به‌نحوی مشابه، در حالت گسسته در زمان، تبدیل فوریه‌های ورودی $X(e^{j\omega})$ و خروجی $Y(e^{j\omega})$ سیستم LTI توسط پاسخ فرکانسی $H(e^{j\omega})$ به هم مربوط می‌شوند:  

$$
Y(e^{j\omega}) = H(e^{j\omega}) X(e^{j\omega}) \quad (6-4)
$$

پس اثر سیستم LTI بر ورودی، تغییر اندازه مولفه‌های فرکانسی آن است. اگر از دید نمایش اندازه - فاز به این اثر بنگریم، جزئیات بیشتری را خواهیم دید. در حالت پیوسته در زمان:  

$$
|Y(j\omega)| = |H(j\omega)| |X(j\omega)| \quad (6-5)
$$

و  

$$
\angle Y(j\omega) = \angle H(j\omega) + \angle X(j\omega) \quad (6-6)
$$

در حالت گسسته در زمان نیز رابطه دقیقاً مشابهی وجود دارد.  

معادله (6-5) نشان می‌دهد که اثر یک سیستم LTI بر اندازه تبدیل فوریه ورودی، تغییر دادن مقیاس آن بر اساس پاسخ فرکانسی است. به همین دلیل $H(j\omega)$ یا $|H(e^{j\omega})|$ را غالباً **بهره سیستم** می‌نامند.  

معادله (6-6) نشان می‌دهد که سیستم LTI فاز ورودی $\angle X(j\omega)$ را با افزودن فاز $\angle H(j\omega)$ به آن تغییر می‌دهد و به همین دلیل $\angle H(j\omega)$ را غالباً **جابجایی فاز سیستم** می‌نامند.  

جابجایی فاز سیستم می‌تواند رابطه نسبی فاز مولفه‌های ورودی را تغییر دهد و منجر به تغییر شدید چهره حوزه زمان ورودی شود، حتی در مواقعی که بهره سیستم در همه فرکانس‌ها ثابت است.  
تغییر اندازه و فاز توسط سیستم LTI می‌تواند مطلوب یا نامطلوب باشد، بسته به این‌که این تغییر چگونه صورت می‌گیرد. 

در حالت دوم، یعنی نامطلوب بودن تغییر، می‌گوییم دامنه و فاز **اعوجاج** یافته‌اند.
در بخش‌های زیر مفاهیم و ابزارهایی برای ملموس کردن این اثرها پی‌ریزی خواهیم کرد.
