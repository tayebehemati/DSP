# مشخصات زمانی و فرکانسی سیگنال‌ها و سیستم‌ها

## ۰-۶ مقدمه  
<div dir="rtl" style="text-align: right;">
توصیف یک سیستم LTI بر حسب پاسخ فرکانسی جایگزینی برای توصیف حوزه زمان سیستم بر حسب کانولوشن است.  
در تحلیل سیستم‌های LTI غالباً استفاده از حوزه فرکانسی مناسب‌تر است زیرا معادلات دیفرانسیل و تفاضلی و عمل کانولوشن در حوزه فرکانس به عملیات جبری تبدیل می‌شوند.  
همچنین مفاهیمی چون فرکانس‌گزینی را در حوزه فرکانس می‌توان راحت تصور کرد.  
<div dir="rtl" style="text-align: right;">
البته در طراحی سیستم معمولاً باید هم ملاحظات حوزه فرکانس منظور شود و هم ملاحظات حوزه زمان.  
برای نمونه همان‌طور که در مثال‌های ۴-۱۸ و ۵-۱۲ به اختصار گفتیم، و در این فصل به تفصیل خواهیم گفت، رفتار نوسانی شدید پاسخ ضربه فیلترهای فرکانس‌گزین ممکن است نامطلوب باشد و مجبور شویم فرکانس‌گزینی را تا حدی فدای دستیابی به مشخصات مطلوب حوزه زمان کنیم.  
<div dir="rtl" style="text-align: right;">
در عمل وضعیت‌های این‌چنین قاعده هستند نه استثنا، زیرا در اکثر کاربردها قیدهایی برای سیستم هم در حوزه زمان و هم در حوزه فرکانس تعیین می‌کنیم که به خواسته‌های متضاد می‌انجامد.  
<div dir="rtl" style="text-align: right;">
پس در طراحی و تحلیل سیستم دانستن روابط بین حوزه فرکانس و حوزه زمان و بده‌بستان‌های موجود مهم است.  
معرفی این مباحث و روابط مهم‌ترین هدف این فصل است.  

<div dir="rtl" style="text-align: right;">

## ۱-۶ نمایش اندازه-فاز تبدیل فوریه  
<div dir="rtl" style="text-align: right;">
تبدیل فوریه در حالت کلی مختلط است و همان‌طور که قبلاً گفته‌ایم می‌توان آن را با بخش‌های حقیقی و موهومی یا اندازه و فازش نشان داد.  
نمایش اندازه - فاز تبدیل فوریه پیوسته در زمان $X(j\omega)$ عبارت است از:  

$$
X(j\omega) = |X(j\omega)| e^{j \angle X(j\omega)} \tag{۶-۱}
$$

<div dir="rtl" style="text-align: right;">
به‌نحوی مشابه، نمایش اندازه - فاز تبدیل فوریه گسسته در زمان $X(j\omega)$ عبارت است از:

$$
X(j\omega) = |X(j\omega)| e^{j \angle X(j\omega)} \tag{۶-۲}
$$

<div dir="rtl" style="text-align: right;">
در بحث زیر راجع به نمایش‌های اندازه - فاز توجه عمده خود را روی حالت پیوسته در زمان متمرکز می‌کنیم.  
نکات اصلی را به حالت گسسته در زمان نیز می‌توان اعمال کرد.  
<div dir="rtl" style="text-align: right;">   
از معادله ترکیب تبدیل فوریه یعنی معادله (۴-۸) چنین بر می‌آید که می‌توان $X(j\omega)$ را تجزیه‌ی سیگنال $x(t)$ به مجموع نمایی‌های مختلط با فرکانس‌های مختلف دانست.  
در حقیقت چنان‌که در بخش ۴-۳-۷ گفتیم $|X(j\omega)|^۲$ را می‌توان طیف چگالی انرژی $x(t)$ دانست.  
یعنی $|X(j\omega)|^۲ \, d\omega / ۲\pi$ را می‌توان انرژی واقع در باند فرکانسی بی‌نهایت کوچک بین $\omega$ و $\omega + d\omega$ فرض کرد.  
پس اندازهٔ $|X(j\omega)|$ محتوای فرکانسی سیگنال را نشان می‌دهد.  
یعنی $|X(j\omega)|$ اطلاعات مربوط به اندازه نسبی نمایی‌های مختلط تشکیل‌دهندهٔ $x(t)$ را در بردارد.  
مثلاً اگر در خارج یک باند فرکانسی کوچک حول صفر داشته باشیم $|X(j\omega)| = ۰$، آنگاه $x(t)$ تنها نوسان‌های فرکانس پایین دارد.
<div dir="rtl" style="text-align: right;">
ولی زاویه فاز $\angle X(j\omega)$ بر اندازهٔ مولفه‌های فرکانسی اثری ندارد، بلکه اطلاعات مربوط به فاز نسبی این نمایی‌ها را نشان می‌دهد.  
رابطهٔ فاز موجود در $\angle X(j\omega)$ اثر زیادی بر طبیعت سیگنال $x(t)$ دارد و نوعاً اطلاعات زیادی در مورد سیگنال دارد.  
در واقع بسته به تابع فاز می‌توانیم سیگنال‌هایی با ظاهر متفاوت به دست آوریم، حتی اگر تابع اندازه تغییر نکند.برای نمونه دوباره مثال شکل ۳-۳ را در نظر بگیرید.  
در اینجا کشتی با جمع آثار سه موج روبرو است، که هر کدام با یک سیگنال سینوسی مدل شده‌اند.  
اگر این سه سینوسی دامنهٔ ثابتی داشته باشند، بسته به فاز نسبی‌شان می‌توانند دامنه بسیار بزرگ یا کوچکی داشته باشند.پس فازها برای کشتی کاملاً مهم‌اند.به عنوان نمونه دیگری از اثر فاز سیگنال زیر را در نظر بگیرید.  

$$
x(t) = ۱ + \frac{۱}{۲} \cos(۲\pi t + \varphi_۱) + \cos(۴\pi t + \varphi_۲) + \frac{۲}{۳} \cos(۶\pi t + \varphi_۳) \tag{۶-۳}
$$

<div dir="rtl" style="text-align: right;">
شکل ۳-۴ این سیگنال را به‌ازای $\varphi_۱ = \varphi_۲ = \varphi_۳ = ۰$ نشان می‌داد.  
در شکل ۱-۶ این $x(t)$ را به‌ازای چند فاز مخالف برای سینوس‌ها نشان داده‌ایم.  
همان‌طور که از شکل برمی‌آید، می‌بینیم که سیگنال حاصل بر حسب مقادیر مختلف فاز نسبی چقدر می‌تواند متفاوت باشد.  
<div dir="rtl" style="text-align: right;">
به‌طور کلی تغییر تابع فاز $X(j\omega)$ به تغییر مشخصات حوزه زمان سیگنال $x(t)$ می‌انجامد.  
در بعضی موارد اعوجاج فاز مهم است و در بعضی موارد نه.  
مثلاً یکی از خواص سیستم شنوایی انسان عدم حساسیت به فاز است.  
اگر تبدیل فوریهٔ سیگنال یک صدای تنها (مثلاً یک مصوت) اعوجاج فاز داشته باشد ولی اندازه‌اش تغییر نکند،  
قابلیت فهم آن تغییر زیادی نمی‌کند، گرچه شکل موج آن در حوزه زمان می‌تواند شکلی کاملاً دگرگون به خود بگیرد.  
هرچند اعوجاج ملایم فاز، نظیر اعوجاج تک‌صداها، قابلیت تشخیص را از بین نمی‌برد،  
ولی اعوجاج فاز شدید گفتار، مسلماً قابلیت تشخیص آن‌ها را از بین می‌برد.

<div dir="rtl" style="text-align: right;">
به عنوان مثالی بارز فرض کنید $x(t)$ سیگنال ضبط شده‌ی بیان یک جمله است. اگر نوار وارونه پخش شود، سیگنال $x(-t)$ به دست می‌آید، و اثر متناظر در حوزه فرکانس منفی شدن فاز است.  

$$
F\{x(-t)\} = X(-j\omega) = |X(j\omega)| e^{-j\angle x(j\omega)}
$$
<div style="text-align: center;">

<p dir="rtl" style="font-size:10px; text-align:center;">
شکل ۱-۶ سیگنال $x(t)$ معادله (۳-۶) به‌ازای چند مقدار مختلف $\varphi_۱$، $\varphi_۲$، و $\varphi_۳$:  
(الف) $\varphi_۱ = \varphi_۲ = \varphi_۳ = ۰$، (ب) $\varphi_۱ = ۱۲ \ rad$، $\varphi_۲ = ۸ \ rad$، $\varphi_۳ = ۴ \ rad$،
</p>
<p dir="rtl" style="font-size:10px; text-align:center;">
(ج) $\varphi_۱ = ۶ \ rad$، $\varphi_۲ = -۲.۷ \ rad$، $\varphi_۳ = ۰.۹۳ \ rad$،  
(د) $\varphi_۱ = ۱.۲ \ rad$، $\varphi_۲ = ۴.۱ \ rad$، $\varphi_۳ = -۷.۰۲ \ rad$.
</p>  

<div dir="rtl" style="text-align: right;">
یعنی اندازه طیف نوار وارونه با طیف سیگنال اصلی برابر است و تنها فاز آن تغییر کرده است. واضح است که این تغییر فاز اثر مهمی بر قابلیت فهم صدای ضبط شده دارد.  
<div dir="rtl" style="text-align: right;">
بررسی تصاویر مثال دیگری است که اهمیت فاز را نشان می‌دهد. همانطور که در فصل ۳ به اختصار گفتیم، تصویر سیاه و سفید را می‌توان با سیگنال $x(t_۱,t_۲)$ نشان داد که $t_۱$ مختصه‌ی افقی نقطه‌ای از تصویر، $t_۲$ مختصه‌ی قائم آن نقطه و $x(t_۱,t_۲)$ روشنی تصویر در $(t_۱,t_۲)$ است.  
تبدیل فوریه $X(j\omega_۱,j\omega_۲)$ تصویر، تجزیه‌ی $x(t_۱,t_۲)$ به صورت نمایی‌های $e^{\omega_۱ t_۱}$ و $e^{\omega_۲ t_۲}$ را نشان می‌دهد که تغییرات مکان $x(t_۱,t_۲)$ در فرکانس‌های مختلف را در بردارد. جنبه‌های مقدماتی تحلیل فوریه دوبعدی در مسائل ۴-۵۳ و ۴-۵۶ بررسی شده‌اند.  
<div dir="rtl" style="text-align: right;">
در دیدن تصاویر بخشی از مهم‌ترین اطلاعات در لبه‌ها و نواحی دارای کنتراست بالا نهفته است. نواحی دارای ماکزیمم و مینیمم روشنی در یک تصویر، محل‌هایی هستند که توابع نمایی مختلط دارای فرکانس‌های متفاوت هم‌فازند.  
بنابراین معقول است که تصور کنیم فاز تبدیل فوریه یک تصویر بخش اعظم اطلاعات راجع به تصویر را در بر داشته باشد و مخصوصاً اطلاعات راجع به لبه‌ها باید در فاز باشد.  
برای بررسی این تصور، شکل ۱-۴ را دوباره در شکل ۶-۲ (الف) نشان داده‌ایم.  
شکل ۶-۲ (ب) اندازه‌ی تبدیل فوریه را نشان می‌دهد.  
در این شکل محور افقی $\omega_۱$ و محور عمودی $\omega_۲$ و روشنی تصویر در نقطه $(\omega_۱,\omega_۲)$ متناسب با اندازه تبدیل $X(j\omega_۱,j\omega_۲)$ تصویر شکل ۶-۲ (الف) است.  
فاز این تبدیل هم به نحوی مشابه در شکل ۶-۲ (ج) نشان داده شده است.  
شکل ۶-۲ (د) نتیجه صفر کردن فاز $X(j\omega_۱,j\omega_۲)$، حفظ اندازه تبدیل و گرفتن عکس تبدیل را نشان می‌دهد.  
در شکل ۶-۲ (ه) اندازه $X(j\omega_۱,j\omega_۲)$ برابر یک قرار داده شده، ولی فاز حفظ شده است.  
سرانجام شکل ۶-۲ (و) تصویر حاصل از گرفتن عکس تبدیل از تابعی با فاز نشان داده شده در شکل ۶-۲ (ج) و اندازه تبدیل تصویری <b>کاملاً متفاوت</b> (یعنی تصویر شکل ۶-۲ (ز)) را نشان می‌دهد.  
این تصاویر به روشنی اهمیت فاز در نمایش تصاویر را نشان می‌دهند.

## ۶-۲ نمایش اندازه - فاز پاسخ فرکانسی سیستم‌های LTI  
<div dir="rtl" style="text-align: right;">
طبق خاصیت کانولوشن، تبدیل فوریه پیوسته در زمان $Y(jω)$ خروجی یک سیستم LTI و تبدیل $X(jω)$ ورودی آن به صورت زیر به هم مربوط می‌شوند:  

$$
Y(jω) = H(jω)X(jω)
$$

<div dir="rtl" style="text-align: right;">
که در آن $H(jω)$ پاسخ فرکانسی سیستم؛ یعنی تبدیل فوریه پاسخ ضربه سیستم است. به نحوی مشابه، در حالت گسسته در زمان، تبدیل فوریه‌های ورودی $X(e^{jω})$ و خروجی $Y(e^{jω})$ سیستم LTI توسط پاسخ فرکانسی $H(e^{jω})$ به هم مربوط می‌شوند:

$$
Y(e^{jω}) = H(e^{jω}) X(e^{jω}) \tag{۶-۴}
$$
<div dir="rtl" style="text-align: right;">
پس اثر سیستم LTI بر ورودی، تغییر اندازه مولفه‌های فرکانسی آن است. اگر از دید نمایش اندازه - فاز به این اثر بنگریم، جزئیات بیشتری را خواهیم دید. در حالت پیوسته در زمان:

$$
|Y(j\omega)|=|H(j\omega)| |X(j\omega)|\tag{۶-۵}
$$
<div dir="rtl" style="text-align: right;">
و

$$
\angle Y(j\omega)=\angle H(j\omega)+\angle X(j\omega)\tag{۶-۶}
$$
<div dir="rtl" style="text-align: right;">
در حالت گسسته در زمان نیز رابطه دقیقاً مشابهی وجود دارد. معادله (۶-۵) نشان می‌دهد که اثر یک سیستم LTI بر اندازه تبدیل فوریه ورودی، تغییر دادن مقیاس آن بر اساس پاسخ فرکانسی است. به همین دلیل $H(jω)$ [یا $|H(e^{jω})|$] را غالباً <b>بهره</b> سیستم می‌نامند. معادله (۶-۶) نشان می‌دهد که سیستم LTI فاز ورودی $\angle X(jω)$ را با افزودن فاز $\angle H(jω)$ به آن تغییر می‌دهد و به همین دلیل $\angle H(jω)$ را غالباً جابجایی فاز سیستم می‌نامند. <b>جابه جایی فاز</b> سیستم می‌تواند رابطه نسبی فاز مولفه‌های ورودی را تغییر دهد و منجر به تغییر شدید چهره حوزه زمان ورودی شود، حتی در مواقعی که بهره سیستم در همه فرکانس‌ها ثابت است. تغییر اندازه و فاز توسط سیستم LTI می‌تواند مطلوب یا نامطلوب باشد، بسته به این‌که این تغییر چگونه صورت می‌گیرد. در حالت دوم، یعنی نامطلوب بودن تغییر، می‌گوییم دامنه و فاز <b>اعوجاج</b> یافته‌اند. در بخش‌های زیر مفاهیم و ابزارهایی برای ملموس کردن این اثرها پی‌ریزی خواهیم کرد.
</div>

## ۱-۲-۶ فاز خطی و غیر خطی
<div dir="rtl" style="text-align: right;">
اگر تغییر فاز تابعی خطی از $\omega$ باشد، اثر آن تعبیر سرراستی در حوزه زمان دارد. سیستم LTI پیوسته در زمانی با پاسخ فرکانسی زیر را در نظر بگیرید:

$$
H(j\omega) = e^{-j\omega t_۰} \tag{۶-۷}
$$
<div dir="rtl" style="text-align: right;">
پس بهره سیستم یک و فاز آن خطی است، یعنی

$$
|H(j\omega)| = ۱ \quad \angle H(j\omega) = -\omega t_۰ \tag{۶-۸}
$$
<div dir="rtl" style="text-align: right;">
در مثال ۴-۱۵ نشان دادیم که در سیستمی با این پاسخ فرکانسی، خروجی جابجا شده زمانی ورودی است، یعنی

$$
y(t) = x(t - t_۰) \tag{۶-۹}
$$
<div dir="rtl" style="text-align: right;">
در حالت گسسته در زمان، فاز خطی اثری مشابه حالت پیوسته در زمان دارد، به شرطی که شیب فاز خطی عدد صحیح باشد. از مثال ۵-۱۱ به یاد داریم که سیستم LTI دارای پاسخ فرکانسی $e^{-j\omega n_۰}$ تابع فاز خطی دارد و خروجی آن جابجا شده ورودی آن است، یعنی $y[n] = x[n - n_۰]$.
پس فاز خطی دارای شیب صحیح با جابجایی$x[n]$ به اندازه آن عدد صحیح متناظر است. اگر شیب فاز عدد صحیح نباشد، اثری که در حوزه زمان ایجاد می‌شود پیچیده‌تر است و در فصل ۷ بخش ۷-۵ مورد بحث قرار خواهد گرفت.
به طور حسی می‌توان گفت که پوش مقادیر سیگنال جابجا می‌شود ولی خود مقادیر ممکن است تغییر کنند.
<div dir="rtl" style="text-align: right;">
تغییر فاز خطی به تغییر ساده و ملموسی در حوزه زمان منجر می‌شود، ولی اگر ورودی در معرض تغییر فاز غیر خطی قرار گیرد، مولفه‌های مختلف ورودی به نحوی تغییر می‌کنند که فازهای نسبی‌شان عوض می‌شود. با برنهادن این نمایی‌ها سیگنالی به دست می‌آید که منظری کاملاً متفاوت نسبت به سیگنال ورودی دارد.
شکل ۶-۳ این مطلب را برای حالت پیوسته در زمان نشان می‌دهد. شکل ۶-۳ (الف) سیگنالی را نشان می‌دهد که به سه سیستم متفاوت اعمال شده است. شکل ۶-۳ (ب) خروجی سیستمی با پاسخ فرکانسی 
$H_۱(j\omega) = e^{-j\omega n_۰}$
به این ورودی است که تنها یک تاخیر $t_۰$ ثانیه‌ای دارد. در شکل ۶-۳ (ج) خروجی سیستمی با بهره واحد و تابع فاز غیر خطی به این ورودی نشان داده شده است. یعنی پاسخ فرکانسی سیستم به صورت زیر است

$$
H_۲(j\omega) = e^{j\angle H_۲(j\omega)} \tag{۶-۱۰}
$$
<div dir="rtl" style="text-align: right;">
که در آن $\angle H_۲(j\omega)$ تابعی غیر خطی از $\omega$ است.
شکل ۶-۳ (د) خروجی سیستم دیگری با فاز غیر خطی را نشان می‌دهد. این سیستم با افزودن یک فاز خطی به سیستم قبلی به دست آمده است، یعنی

$$
H_۳(j\omega) = H_۲(j\omega) \cdot e^{-j\omega n_۰} \tag{۶-۱۱}
$$
<div dir="rtl" style="text-align: right;">
پس خروجی شکل ۶-۳ (د) را می‌توان پاسخ اتصال متوالی سیستم $H_۲(j\omega)$ و سیستمی با فاز خطی دانست، به نحوی که سیگنال‌های شکل ۶-۳ (ج) و (د) تنها جابجا شده یکدیگر باشند.
<div dir="rtl" style="text-align: right;">
در شکل ۶-۴ اثر فاز خطی و غیر خطی در حالت گسسته در زمان نشان داده شده است. باز هم شکل ۶-۴ (الف) ورودی اعمال شده به سه سیستم LTI است که همگی بهره‌ای برابر یک دارند (یعنی $|H(e^{j\omega})| = ۱$).
سیگنال‌های دیگر شکل ۶-۴ خروجی‌های این سه سیستم هستند. برای شکل ۶-۴ (ب)، سیستم فاز خطی با شیب ۵ دارد. بنابراین خروجی همان ورودی با ۵ واحد تأخیر است.
جابجایی فاز سیستم‌های متناظر با شکل‌های ۶-۴ (ج) و (د) غیر خطی است ولی اختلاف‌شان در یک فاز خطی دارای شیب صحیح است، بنابراین سیگنال‌های شکل‌های ۶-۴ (ج) و (د) تاخیر یافته یکدیگرند.
<div dir="rtl" style="text-align: right;">
توجه کنید که تمام سیستم‌های در نظر گرفته شده در شکل‌های ۶-۳ و ۶-۴ بهره‌ای واحد دارند، یعنی اندازه تبدیل فوریه ورودی بدون تغییر در خروجی ظاهر می‌شود. به همین <b>تمام گذر</b> می‌نامند.

</div>

<p dir="rtl" style="font-size:10px; text-align:center;">  
شکل ۳-۶ سیگنال پیوسته در زمانی که به عنوان ورودی چند سیستم با پاسخ فرکانسی دارای بهره واحد به کار رفته است؛ (الف) خروجی سیستم دارای فاز خطی؛ 
</p>  
<p dir="rtl" style="font-size:10px; text-align:center;">
(ب) خروجی سیستم دارای فاز غیرخطی؛ (ج) پاسخ سیستمی با همان تابع فاز سیستم (ج) به اضافه یک فاز خطی (د).

شکل ۶-۴ (الف)  سیگنال گسته در زمانی که به  چند سیستم با اندازه پاسخ  فرکانی یک اعمال شده  است ؛ (ب) پاسخ سیستم  دارای ناز خطی با شیب ۵-؛

<div dir="rtl" style="text-align: right;">
مشخصات یک سیستم تمام‌گذر به طور کامل توسط جابجایی فاز آن تعیین می‌شود. البته در حالت کلی سیستم‌های LTI هم اندازه تبدیل فوریه را با بهره $|H(j\omega)|$ یا $|H(e^{j\omega})|$ تغییر می‌دهند و هم فاز آن را، و این تغییر فاز می‌تواند خطی یا غیر خطی باشد.

```python
import numpy as np
import matplotlib.pyplot as plt

# تعریف محور زمان
t = np.linspace(-10, 50, 1000)

# سیگنال الف
x_a = np.exp(-((t) ** 2) / 10)

# سیگنال ب
x_b = np.exp(-((t - 10) ** 2) / 10)

# سیگنال ج
x_c = np.zeros_like(t)
mask_c = (t >= 0) & (t <= 30)
x_c[mask_c] = np.sin(np.pi * t[mask_c] / 15)

# سیگنال د
x_d = np.zeros_like(t)
mask_d = (t >= 10) & (t <= 40)
x_d[mask_d] = np.sin(np.pi * (t[mask_d] - 10) / 15)

# رسم نمودارها
fig, axs = plt.subplots(2, 2, figsize=(10, 6), layout='constrained')

titles = ['(الف)', '(ب)', '(ج)', '(د)']
signals = [x_a, x_b, x_c, x_d]

for i, ax in enumerate(axs.flat):
    ax.plot(t, signals[i], color='black')
    ax.set_title(titles[i], fontsize=12)
    ax.set_xlabel('زمان (t)')
    ax.grid(True)

plt.suptitle('نمایش چهار سیگنال مختلف', fontsize=14)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-10, 21)

def signal_a(n):
    return np.where((n >= -3) & (n <= 3), 1 - (np.abs(n) / 4), 0)

def signal_b(n):
    return signal_a(n - 5)

def signal_c(n):
    return np.where(n >= 0, (0.9) ** n, 0) - np.where(n <= 0, (0.9) ** (-n), 0)

def signal_d(n):
    return signal_c(n - 5)

fig, axs = plt.subplots(4, 1, figsize=(6, 8))

signals = [signal_a, signal_b, signal_c, signal_d]
titles = ["نمودار (الف)", "نمودار (ب)", "نمودار (ج)", "نمودار (د)"]

for i in range(4):
    axs[i].stem(n, signals[i](n), basefmt=" ")
    axs[i].axhline(0, color='black', linewidth=1)
    axs[i].set_title(titles[i])
    axs[i].set_xlabel("n")

plt.tight_layout()

from IPython.display import display, HTML
display(HTML("<style>.output_png {display: flex; justify-content: center;}</style>"))

plt.show()
```

## ۲-۲-۶ تاخیر گروه
<div dir="rtl" style="text-align: right;">
همان‌طور که در بخش ۶-۲-۱ گفتیم، سیستم‌هایی دارای تغییر فاز خطی این ویژگی را دارند که اثرشان را می‌توان با یک جابجایی زمانی ساده بیان کرد. در واقع طبق معادلات (۶-۸) و (۶-۹)، شیب فاز اندازه جابجایی زمانی را نشان می‌دهد. در حالت پیوسته در زمان، اگر  $\angle H(j\omega) = -\omega t_۰$  
آنگاه جابجایی به اندازه $-t_۰$ است، یعنی تاخیری به اندازه $t_۰$. 
در حالت گسسته در زمان $\angle H(e^{j\omega}) =-\omega n_۰$ با تاخیر $n_۰$ متناظر است.
<div dir="rtl" style="text-align: right;">
مفهوم تاخیر را می‌توان به طور طبیعی تعمیم داد تا فازهای غیر خطی را نیز شامل شود. فرض کنید بخواهیم اثر فاز یک سیستم LTI پیوسته در زمان را بر یک ورودی باریک‌باند بررسی کنیم. منظور از سیگنال باریک‌باند، سیگنالی است که تبدیل فوریه‌اش در خارج باند باریکی حول فرکانس $\omega = \omega_۰$ صفر یا ناچیز است.  
اگر باند سیگنال خیلی باریک باشد می‌توانیم فاز سیستم را در این باند با تعریف خطی زیر بیان کنیم  

$$
\angle H(j\omega) \approx -\varphi - \omega \alpha \tag{۶-۱۲}
$$  

<div dir="rtl" style="text-align: right;">
پس 
    
$$
Y(j\omega) \approx X(j\omega) |H(j\omega)| e^{-j\varphi} e^{-j\omega \alpha} \tag{۶-۱۳}
$$  
 
<div dir="rtl" style="text-align: right;">
اثر تقریبی سیستم بر این سیگنال باریک‌باند، تغییر اندازه با ضریب $|H(j\omega)|$، ضرب در عامل مختلط ثابت، و ضرب در فاز خطی $e^{-j\omega \alpha}$ است که با تاخیر $\alpha$ ثانیه‌ای متناظر است.  
این تاخیر زمانی را <b>تاخیر گروه</b> در $\omega = \omega_۰$ می‌نامند، زیرا تاخیر مشترکی را که فرکانس‌های باند حول فرکانس مرکزی $\omega = \omega_۰$ پیدا می‌کنند، نشان می‌دهد.  

<div dir="rtl" style="text-align: right;">
تاخیر گروه در هر فرکانسی با منفی شیب فاز در آن فرکانس برابر است، یعنی تاخیر گروه به صورت زیر تعریف می‌شود  

$$
\tau(\omega) = -\frac{d}{d\omega} \{ \angle H(j\omega) \} \tag{۶-۱۴}
$$  

<div dir="rtl" style="text-align: right;">
مفهوم تاخیر گروه را می‌توان در مورد سیستم‌های گسسته در زمان نیز به کار برد. در مثال زیر اثر تاخیر گروه غیر ثابت بر یک سیگنال را در نظر می‌گیریم.  

---

### مثال ۶-۱  
<div dir="rtl" style="text-align: right;">
پاسخ ضربه یک سیستم تمام‌گذر با تاخیر گروه متغیر با فرکانس را در نظر بگیرید.  
پاسخ فرکانسی $H(j\omega)$ را حاصل‌ضرب سه عامل فرض می‌کنیم، یعنی  

$$
H(j\omega) = \prod_{i=۱}^{۳} H_i(j\omega)
$$  
<div dir="rtl" style="text-align: right;">
که در آن:  
    
$$
H_i (j\omega) = \frac{۱ + (j\omega / \omega_i )^۲ - ۲j\zeta_i (\omega / \omega_i)}{۱ + (j\omega / \omega_i )^۲ + ۲j\zeta_i (\omega / \omega_i)} \tag{۶-۱۵}
$$

 $$
\left\{
\begin{array}{l}
\omega_۱ = ۳۱۵ \quad \text{rad/sec} \quad \text{و} \quad \zeta_۱ = ۰٫۰۶۶ \\
\omega_۲ = ۹۴۳ \quad \text{rad/sec} \quad \text{و} \quad \zeta_۲ = ۰٫۰۳۳ \\
\omega_۳ = ۱۸۸۸\quad \text{rad/sec} \quad \text{و} \quad \zeta_۳ = ۰٫۰۵۸
\end{array}
\right.
$$


<div dir="rtl" style="text-align: right;">
غالباً بهتر است فرکانس $\omega_i$ با واحد رادیان بر ثانیه را بر حسب فرکانس $f_i$ با واحد هرتز بیان کنیم:  

$$
\omega_i = ۲\pi f_i
$$  
<div dir="rtl" style="text-align: right;">
در اینجا داریم:
    
$$
f_۱ \approx ۵۰ \text{Hz}
$$  
$$
f_۲ \approx ۱۵۰ \text{Hz}
$$  
$$
f_۳ \approx ۳۰۰ \text{Hz}
$$  
<div dir="rtl" style="text-align: right;">
چون صورت هر عامل $H_i(j\omega)$ مزدوج مختلط مخرج آن است، $|H_i(j\omega)| = ۱$. پس می‌توانیم نتیجه بگیریم که

$$
H(j\omega) =۱
$$ 
<div dir="rtl" style="text-align: right;">
فاز هر عامل $H_i(j\omega)$ را می‌توان از معادله (۶-۱۵) به دست آورد.  

$$
\angle H_i (jω)=-۲αρχταv \left[ \frac {۲ζ_i (ω∕ω_i )}{۱-(ω∕ω_i )^۲ } \right]
$$

<div dir="rtl" style="text-align: right;">
و

$$
\angle H(j\omega) = \sum_{i=۱}^{۳} \angle H_i (j\omega)
$$

<div dir="rtl" style="text-align: right;">
اگر مقدار $\angle H(j\omega)$ را در فاصله $-\pi$ تا $\pi$ بگیریم، تابع فاز اصلی را مطابق شکل ۶-۵ (الف) به دست می‌آوریم؛ این شکل فاز را بر حسب فرکانس دارای واحد هرتز نشان می‌دهد.  
توجه کنید که این تابع در بعضی فرکانس‌ها ناپیوستگی‌هایی با اندازه $۲\pi$ دارد، یعنی تابع فاز در این فرکانس‌ها مشتق‌پذیر نیست. ولی با افزودن مضارب صحیح $۲\pi$ به فاز، پاسخ فرکانسی تغییر نمی‌کند.  
با افزودن مضارب صحیح مناسب $۲\pi$ به بخش‌های مختلف فاز، فاز باز شده شکل ۶-۵ (ب) را به دست می‌آوریم. اکنون می‌توانیم تاخیر گروه را بر حسب فرکانس به صورت زیر حساب کنیم. 

$$
\tau(\omega) = -\frac{d}{d\omega} \left\{ \angle [H(j\omega)] \right\}
$$
 

<div dir="rtl" style="text-align: right;">
که در آن $\angle [H(jω)]$ تابع باز شده فاز $H(jω)$ را نشان می‌دهد. شکل شکل ۶-۵ (ج) $τ(ω)$ را نشان می‌دهد. توجه کنید که فرکانس‌های حول ۵۰Hz تاخیری بیشتر از فرکانس‌های حول ۱۵۰Hz با ۳۰۰Hz پیدا می‌کنند. اثر این تاخیر گروه غیر ثابت را می‌توان به طور کیفی در پاسخ ضربه ]شکل ۶-۵ (د)[ سیستم LTI نیز مشاهده کرد. می‌دانیم که $F\{δ(t)\} = ۱$ . مولفه‌های فرکانسی ضربه چنان قرار می‌گیرند که ترکیبشان یک ضربه دست دهد، و ضربه تابعی به شدت محلی است. چون سیستم تمام گذر تاخیر گروه غیر ثابت دارد، فرکانس‌های مختلف ورودی تاخیرهای متفاوتی می‌یابند. این پدیده را پخش می‌نامند. در مثال فعلی بزرگ‌ترین تاخیر گروه در ۵۰Hz رخ می‌دهد. پس بخش انتهایی پاسخ ضربه با فرکانسی حول ۵۰Hz نوسان می‌کند. شکل ۶-۵ (و) به وضوح این پدیده را نشان می‌دهد.
</div>

         alt="شکل ۶-۵"
         style="display: block; margin: auto; width: 50%;">
</div>

<p dir="rtl" style="font-size:10px; text-align:center;">
شکل ۵-۶ فاز، تأخیر گروه و پاسخ ضربه سیستم تمام‌گذر. سیستم تمام‌گذر مثال ۶-۱؛ (الف) فاز اصلی؛ (ب) فاز باز شده؛ (ج) تأخیر گروه؛ (د) پاسخ ضربه. تمام این کمیت‌ها بر حسب فرکانس هرتزی رسم شده‌اند.
</p>

### مثال ۶-۲  

<div dir="rtl" style="text-align: right;">
تاخیر گروه غیر ثابت از عامل‌هایی است که در شبکه‌های مخابراتی قطع و وصلی با اهمیت شمرده می‌شود. شرکت AT&T/Bell نتایج بررسی خود در تمام فلات آمریکا راجع به تاخیر گروه مخابره‌های مختلف گزارش کرده است. (۱) شکل ۶-۶ برخی نتایج این بررسی را برای دو دسته مخابره این گونه نشان می‌دهد. آنچه در منحنی‌های شکل ۶-۶ (الف) نشان داده شده، در واقع بخش غیر ثابت تاخیر گروه است. یعنی تاخیر ثابتی برابر با مینیمم تاخیر در فرکانس‌های مختلف از تاخیر گروه کم شده و تفاضل حاصل در شکل ۶-۶ (الف) نشان داده شده است. بنابراین هر منحنی شکل ۶-۶ (الف) تاخیر اضافی مولفه‌های فرکانسی مختلف هر دسته را نشان می‌دهد. منحنی‌ها با کلمه‌های short و medium متمایز شده‌اند که به ترتیب مخابرات راه نزدیک (۰ تا ۱۸۰ مایل هوایی) و راه متوسط (۱۸۰ تا ۷۲۵ مایل هوایی) را نشان می‌دهد. می‌بینیم که تاخیر گروه در ۱۷۰۰Hz مینیمم است و در دو طرف این فرکانس به طور یکنوا افزایش می‌یابد.

<p dir="rtl" style="font-size:10px; text-align:center;">
شکل ۶-۶ (الف) بخش غیر ثابت تأخیر گروه، و (ب) اندازه پاسخ فرکانسی تلفن‌های راه نزدیک و راه متوسط [گرفته شده از مقاله <i>Duffy</i> و <i>Thatcher</i>]. این منحنی‌ها بر حسب فرکانس رسم شده‌اند و فرکانس با واحد هرتز مشخص شده است.
</p>
<p dir="rtl" style="font-size:10px; text-align:center;">
دامنه پاسخ فرکانسی نیز مطابق معمول بر حسب دسیبل است، یعنی $۲۰ \log_{۱۰} \left( | H(j\omega) | \right)$ 
رسم شده است. این گونه منحنی‌ها را در بخش ۶-۲-۳ به تفصیل بررسی می‌کنیم.
</p>

<div dir="rtl" style="text-align: right;">
با ترکیب مشخصات تاخیر گروه شکل ۶-۶ (الف) و پاسخ فرکانسی شکل ۶-۶ (ب)، پاسخ ضربه‌های شکل ۶-۷ را به دست می‌آوریم.  
پاسخ ضربه شکل ۶-۷ (الف) به دسته راه نزدیک تعلق دارد.  
مؤلفه‌های بسیار کم‌فرکانس و بسیار پُرفرکانس زودتر از مؤلفه‌های فرکانس متوسط بروز می‌کنند.  
این ویژگی با آنچه شکل ۶-۶ (الف) نشان می‌دهد تطابق دارد.  
شکل ۶-۷ (ب) همین پدیده را برای تلفن‌های راه متوسط نشان می‌دهد.

(الف) پاسخ ضربه‌ متناظر با تلفن‌های راه نزدیک؛  
(ب) پاسخ ضربه‌ متناظر با تلفن‌های راه دور.

```python
from urllib.request import urlopen
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def read_image_from_url(url):
    img = Image.open(urlopen(url)).convert('L')
    img = img.resize((512, 512))
    return np.array(img)

# لینک‌ها
url_alpha = 'https://raw.githubusercontent.com/tayebehemati/DSP/c50e285a5201e5c6405a9e711b7ae3d796894724/IMG_0149.jpeg'
url_z     = 'https://raw.githubusercontent.com/tayebehemati/DSP/0acf61f3be1b43cb8cd97e905c80dcf0526a1bdf/IMG_0156.jpeg'

# خواندن تصاویر
img  = read_image_from_url(url_alpha)
img2 = read_image_from_url(url_z)

# تبدیل فوریه
f       = np.fft.fft2(img)
fshift  = np.fft.fftshift(f)
magnitude = np.abs(fshift)
phase     = np.angle(fshift)

# (ج) فقط فاز
only_phase = np.abs(
    np.fft.ifft2(np.fft.ifftshift(np.exp(1j * phase)))
)

# (د) فقط اندازه (فاز صفر)
only_magnitude = np.abs(
    np.fft.ifft2(np.fft.ifftshift(magnitude))
)

# (ه) اندازه=۱، فاز=اصل
recon_mag1_phase = np.abs(
    np.fft.ifft2(np.fft.ifftshift(np.exp(1j * phase)))
)

# (و) فاز (الف) + اندازه (ز)
f2       = np.fft.fft2(img2)
f2shift  = np.fft.fftshift(f2)
magnitude2 = np.abs(f2shift)
recon_mix  = np.abs(
    np.fft.ifft2(np.fft.ifftshift(magnitude2 * np.exp(1j * phase)))
)

# نمایش
titles = ['(الف) اصلی', '(ب) اندازه فوریه', '(ج) فاز فوریه',
          '(د) فقط اندازه', '(ه) اندازه=۱, فاز', '(و) اندازه (ز)+فاز', '(ز) تصویر دوم']
images = [img, np.log1p(magnitude), phase,
          only_magnitude, recon_mag1_phase, recon_mix, img2]

plt.figure(figsize=(10,10))
for i in range(7):
    plt.subplot(4,2,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(12, 6))

# --- نمودار 1 ---
f = np.linspace(0.1, 400, 1000)
T = 100
phase = np.pi * (f % T) / T
phi1 = np.cos(phase) / (np.sin(phase) + 1e-10)
phi1 = np.clip(phi1, -3.9, 3.9)
axs[0, 0].plot(f, phi1, 'k', linewidth=2)
axs[0, 0].set_xlim(0, 400)
axs[0, 0].set_ylim(-4, 4)
axs[0, 0].set_xlabel('فرکانس (Hz)')
axs[0, 0].set_ylabel('فاز (rad)')
axs[0, 0].grid(True)

# --- نمودار 2 ---
f = np.linspace(0, 400, 1000)
phi2 = -5 * np.tanh((f - 50) / 10) - 5 * np.tanh((f - 150) / 10) - 5 * np.tanh((f - 300) / 15)
axs[0, 1].plot(f, phi2, color='black')
axs[0, 1].set_xlabel('فرکانس (Hz)')
axs[0, 1].set_ylabel('فاز (rad)')
axs[0, 1].grid(True)

# --- نمودار 3 ---
peak1 = 0.08 * np.exp(-((f - 50) ** 2) / (2 * 5 ** 2))
peak2 = 0.06 * np.exp(-((f - 150) ** 2) / (2 * 7 ** 2))
peak3 = 0.03 * np.exp(-((f - 300) ** 2) / (2 * 12 ** 2))
signal = peak1 + peak2 + peak3
axs[1, 0].plot(f, signal, color='black')
axs[1, 0].set_xlabel('فرکانس (Hz)')
axs[1, 0].set_ylabel('گذر زمان (sec)')
axs[1, 0].grid(True)

# --- نمودار 4 ---
t = np.linspace(0, 0.2, 1000)
fast = 400 * np.exp(-80 * t) * np.sin(150 * 2 * np.pi * t)
slow = 30 * np.exp(-5 * t) * np.sin(20 * 2 * np.pi * t)
combo = fast + slow
axs[1, 1].plot(t, combo, color='black')
axs[1, 1].set_xlabel('زمان (sec)')
axs[1, 1].set_ylabel('دامنه')
axs[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# محور فرکانس
f = np.linspace(0, 3600, 500)

# ================= نمودار اول =================
# مدل‌سازی دو منحنی مشابه
curve1_a = -((f - 1800)**2) / (1e5) + 0
curve2_a = -((f - 1800)**2) / (1e5) - 2  
curve1_a = np.clip(curve1_a, -30, 5)
curve2_a = np.clip(curve2_a, -30, 5)

# ================= نمودار دوم =================
curve1_b = ((f - 1800)**2) / (1e3) + 500
curve2_b = ((f - 1800)**2) / (1e3) + 1000  

# ================= رسم کنار هم =================
plt.figure(figsize=(12, 4))  # افزایش عرض شکل برای نمایش افقی

# --- subplot اول ---
plt.subplot(1, 2, 1)
plt.plot(f, curve1_a, color='black')
plt.plot(f, curve2_a, color='black')
plt.xlabel('Frequency in (Hz)', fontsize=12)
plt.ylabel('20 log10 |H(w)|', fontsize=12)
plt.title('نمودار اول', fontsize=14)
plt.grid(True)
plt.ylim(-30, 5)
plt.xlim(0, 3600)
plt.text(400, -3, 'Short', fontsize=10)
plt.text(1000, -10, 'Medium', fontsize=10)
plt.text(2900, -5, 'Short', fontsize=10)
plt.text(2300, -10, 'Medium', fontsize=10)

# --- subplot دوم ---
plt.subplot(1, 2, 2)
plt.plot(f, curve1_b, color='black')
plt.plot(f, curve2_b, color='black')
plt.xlabel('Frequency in (Hz)', fontsize=12)
plt.ylabel('Nonconstant Gi (microsecond)', fontsize=12)
plt.title('نمودار دوم', fontsize=14)
plt.grid(True)
plt.ylim(0, 7000)
plt.xlim(0, 3600)
plt.text(400, 4500, 'Medium', fontsize=10)
plt.text(1000, 1000, 'Short', fontsize=10)
plt.text(2900, 4500, 'Medium', fontsize=10)
plt.text(3300, 2200, 'Short', fontsize=10)

# --- نمایش نهایی ---
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# تعریف محور زمان
t = np.linspace(0, 10, 1000)

# ---------- سیگنال اول (الف) ----------
signal1 = np.sin(10 * t) * np.exp(-t) * 0.5
signal1 += 0.1

# ---------- سیگنال دوم (ب) ----------
signal2 = np.sin(15 * t) * np.exp(-0.5 * t) * 0.5
signal2 += 0.05

# ---------- رسم هر دو سیگنال در کنار هم ----------
plt.figure(figsize=(12, 3))

# نمودار (الف)
plt.subplot(1, 2, 1)
plt.plot(t, signal1, color='black')
plt.xlabel('زمان (msec)', fontsize=12)
plt.title('(الف)', fontsize=14)
plt.grid(True)
plt.xlim(0, 10)
plt.ylim(-0.5, 0.6)

# نمودار (ب)
plt.subplot(1, 2, 2)
plt.plot(t, signal2, color='black')
plt.xlabel('زمان (msec)', fontsize=12)
plt.title('(ب)', fontsize=14)
plt.grid(True)
plt.xlim(0, 10)
plt.ylim(-0.5, 0.6)

# نمایش نهایی
plt.tight_layout()
plt.show()
```
