<div dir="rtl" style="text-align: right;">
به بیان دیگر، برای سیستم درجه اول، مجانی‌های لگاریتم دامنه در فرکانس‌های پایین و بالا، خط راست‌اند. مجانب فرکانس پایین [معادله (۶-۲۶)] خط $۰dB$ و مجانب فرکانس بالا [معادله (۶-۲۷)] خطی است که به ازای هر دهه افزایش (یعنی ۱۰ برابر شدن آن)، $|H(j\omega)|$ به اندازه $۲۰dB$ افت می‌کند. این مجانب را گاهی مجانب «$۲۰dB$ بر دهه» می‌نامند.
</div>
<div dir="rtl" style="text-align: right;">
توجه کنید که دو تقریب مجانبی بیان شده با معادلات (۶-۲۶) و (۶-۲۷) در نقطه  
$\log_{۱۰}(j\omega) = -\log_{۱۰}(\tau)$  
با $\omega = \dfrac{۱}{\tau}$ مساوی‌اند. از لحاظ ترسیمی، دو مجانب همدیگر را در نقطه $\omega = \dfrac{۱}{\tau}$ قطع می‌کنند.  
تقریب مجانبی لگاریتم دامنه با دو خط راست با توجه به این مطلب به دست می‌آید.  
تقریب $۲۰\log_{۱۰}|H(j\omega)|$ با صفر در $\omega \ll \dfrac{۱}{\tau}$ و با معادله (۶-۲۷) در $\omega \gg \dfrac{۱}{\tau}$ داده می‌شود.  
این تقریب نیز در شکل ۶-۲۰ (با خط چین) نشان داده شده است.
نقطه‌ای که در آن شیب تقریب تغییر می‌کند، دقیقاً $\omega = \dfrac{۱}{\tau}$ است که به همین دلیل آن را <b>فرکانس شکست</b> می‌نامند.  
همچنین توجه کنید که در $\omega = \dfrac{۱}{\tau}$، دو جملهٔ آرگومان لگاریتم معادله (۶-۲۵) برابرند.  
پس در این نقطه مقدار واقعی اندازه عبارت است از
</div>

$$
۲۰\log_{۱۰}|H(j\dfrac{۱}{\tau})| = -۱۰\log_{۱۰}(۲) \approx -۳\,dB
$$

<div dir="rtl" style="text-align: right;">
به همین خاطر گاهی نقطه $\omega = \dfrac{۱}{\tau}$ را نقطه dB۳ می‌نامند.  
شکل نشان می‌دهد که نمودار بودی تنها در حوالی فرکانس شکست با مجانب خط راست تفاوت زیادی دارد.  
بنابراین برای ترسیم دقیق‌تر نمودار بودی، تنها باید مجانب‌ها را نزدیک فرکانس شکست اصلاح کنیم.
برای $\angle H(j\omega)$ نیز می‌توان تقریب خط راست مفیدی به دست آورد
</div>

$$
\angle H(j\omega) = -\tan^{-۱}(\omega\tau) \approx 
\begin{cases}
۰ & \omega \leq \dfrac{۰٫۱}{\tau} \\
-\dfrac{\pi}{۴}[\log_{۱۰}(\omega\tau)+۱] & \dfrac{۰٫۱}{\tau} \leq \omega \leq \dfrac{۱۰}{\tau} \\
-\dfrac{\pi}{۲} & \omega \geq \dfrac{۱۰}{\tau}
\end{cases}
$$
<div dir="rtl" style="text-align: right;">
توجه کنید که در بازه $\dfrac{۰٫۱}{\tau} \leq \omega \leq \dfrac{۱۰}{\tau}$،  
مقدار این تقریب (از صفر تا $-\dfrac{\pi}{۲}$) به‌صورت خطی کاهش می‌یابد.  
مجانب واقعی $\angle H(j\omega)$ به ازای $\omega \ll \dfrac{۱}{\tau}$ برابر $۰$  
و به ازای $\omega \gg \dfrac{۱}{\tau}$ برابر $-\dfrac{\pi}{۲}$ است.  
همچنین تقریب در فرکانس شکست $\omega = \dfrac{۱}{\tau}$ با مقدار واقعی برابر است. در این نقطه داریم:
</div>

$$
 \angle H(j\omega)(j\dfrac{۱}{\tau}) = -\dfrac{\pi}{۴} \tag{۴-۱۶۸}
$$

<div dir="rtl" style="text-align: right;">
تقریب مجانبی فاز نیز در شکل ۶-۲۰ رسم شده است و این شکل، چگونگی اصلاح مجانب‌های راست برای دقیق‌تر شدن منحنی $\angle H(j\omega)$ را نیز نشان می‌دهد.
با توجه به این سیستم مرتبه اول، بار دیگر می‌توان رابطه معکوس بین حوزه‌های زمان و فرکانس را مشاهده کرد.  
هرچه $\tau$ را کوچکتر کنیم، پاسخ زمانی سیستم سریع‌تر می‌شود (یعنی $h(t)$ به طرف مبدأ فشرده‌تر، و زمان صعود پاسخ پله کوچکتر می‌شود)  
و به‌طور هم‌زمان فرکانس شکست بزرگ‌تر می‌شود (یعنی $H(j\omega)$ بهتر می‌شود زیرا روی گستره فرکانس بزرگ‌تری $|H(j\omega)| \approx ۱$ است).  
با ضرب پاسخ ضربه در $\tau$ و مقایسه $\tau h(t)$ و $H(j\omega)$ هم می‌توان این نکته را دریافت:
</div>

$$
\tau h(t) = e^{-\dfrac{t}{\tau}} u(t)
$$

$$
H(j\omega) = \dfrac{۱}{j\omega\tau + ۱}
$$

<div dir="rtl" style="text-align: right;">
به این ترتیب $\tau h(t)$ تابعی از $\dfrac{t}{\tau}$ و $H(j\omega)$ تابعی از $\omega\tau$ می‌شود.  
بنابراین تغییر با تغییر مقیاس زمان و فرکانس متناظر است.
</div>
